Para garantir consist√™ncia e compatibilidade com a estrutura interna da fundo, foram realizados os seguintes mapeamentos:

| Coluna Original           | Coluna Nova               | Coment√°rio |

### <span style="color:#AEE5F9;"> Gera√ß√£o do Relat√≥rio Final

Consolida√ß√£o de todas as an√°lises em um relat√≥rio HTML interativo.

```python
dimensoes_analise = {
    'Cedentes': 'CedenteNome',
    'Tipo de Ativo': 'TipoAtivo',
    'Situa√ß√£o': 'Situacao',
    'Pagamento Parcial': 'PagamentoParcial',
    'Flag √Ågio/Des√°gio': '_FlagDesagio'
}
dimensoes_analise = {k: v for k, v in dimensoes_analise.items() if v in df_report.columns}
os.makedirs(output_path, exist_ok=True)
```

```python

#***********************
#* C√ÅLCULO M√âTRICAS
#***********************
print("\n" + "="*80)
print("INICIANDO C√ÅLCULO UNIFICADO DAS M√âTRICAS")
print("="*80)

dias_de_atraso = (data_referencia_relatorio - df_report['DataVencimento']).dt.days
df_report['_ValorVencido_1d'] = df_report.loc[dias_de_atraso >= 1, 'ValorPresente'].fillna(0)
df_report['_ValorVencido_30d'] = df_report.loc[dias_de_atraso >= 30, 'ValorPresente'].fillna(0)
df_report['_ValorVencido_60d'] = df_report.loc[dias_de_atraso >= 60, 'ValorPresente'].fillna(0)
df_report.fillna({'_ValorVencido_1d': 0, '_ValorVencido_30d': 0, '_ValorVencido_60d': 0}, inplace=True)

vp_col_name = 'Valor Presente \n(R$ MM)'
vl_col_name = 'Valor L√≠quido \n(R$ MM)'
col_contratos_venc_perc = f"% Contratos Venc. (>{DIAS_ATRASO_DEFINICAO_VENCIDO}d)"
tabelas_metricas = {}

for nome_analise, coluna in dimensoes_analise.items():
    print(f"Calculando m√©tricas para a dimens√£o: '{nome_analise}'...")
    grouped = df_report.groupby(coluna, observed=True)
    total_contratos_unicos = grouped['CCB'].nunique()
    contratos_vencidos_unicos = df_report[df_report['_ContratoVencido_Flag'] == 1].groupby(coluna, observed=True)['CCB'].nunique()
    df_metricas = pd.DataFrame({'N¬∫ Contratos √önicos': total_contratos_unicos})
    somas_financeiras = grouped[['_ValorLiquido', 'ValorPresente', '_ValorVencido_1d', '_ValorVencido_30d', '_ValorVencido_60d']].sum()
    df_metricas = df_metricas.join(somas_financeiras)
    df_metricas['%PDD'] = (1 - df_metricas['_ValorLiquido'] / df_metricas['ValorPresente']) * 100
    df_metricas[col_contratos_venc_perc] = (contratos_vencidos_unicos.reindex(df_metricas.index).fillna(0) / df_metricas['N¬∫ Contratos √önicos']).fillna(0) * 100
    df_metricas['Venc. 1d / VP'] = (df_metricas['_ValorVencido_1d'] / df_metricas['ValorPresente']) * 100
    df_metricas['Venc. 30d / VP'] = (df_metricas['_ValorVencido_30d'] / df_metricas['ValorPresente']) * 100
    df_metricas['Venc. 60d / VP'] = (df_metricas['_ValorVencido_60d'] / df_metricas['ValorPresente']) * 100
    df_metricas = df_metricas.rename(columns={'ValorPresente': vp_col_name, '_ValorLiquido': vl_col_name, 'N¬∫ Contratos √önicos': 'N¬∫ Contratos'})
    df_metricas[[vp_col_name, vl_col_name]] /= 1e6
    df_metricas = df_metricas.drop(columns=['_ValorLiquido', '_ValorVencido_1d', '_ValorVencido_30d', '_ValorVencido_60d'], errors='ignore')
    tabelas_metricas[nome_analise] = df_metricas
print("C√°lculo unificado de m√©tricas conclu√≠do.")
```

```python

#***********************
#* TICKET M√âDIO
#***********************
tabelas_ticket = {}
for nome_analise, coluna in dimensoes_analise.items():
    df_temp = df_report.dropna(subset=[coluna, 'ValorPresente', 'ValorNominal'])
    if df_temp.empty: continue
    grouped = df_temp.groupby(coluna, observed=True)
    numerador = grouped.apply(lambda g: (g['ValorNominal'] * g['ValorPresente']).sum(), include_groups=False)
    denominador = grouped['ValorPresente'].sum()
    ticket_ponderado = (numerador / denominador).replace([np.inf, -np.inf], 0)
    ticket_ponderado.name = "Ticket Ponderado (R$)"
    tabelas_ticket[nome_analise] = pd.DataFrame(ticket_ponderado)
print("Ticket m√©dio calculado.")
```

#### <span style="color:#FFDAC1;"> C√°lculo da TIR

C√°lculo da TIR (XIRR) para a carteira a vencer. Como n√£o h√° dados de custo, calculamos a TIR Bruta e a TIR L√≠quida de PDD.

```python
#******************
#* TIR com brentq
#*******************
def calculate_xirr(cash_flows, days):
    cash_flows = np.array(cash_flows)
    days = np.array(days)
    def npv(rate):
        if rate <= -1: return float('inf')
        with np.errstate(divide='ignore', over='ignore'):
            return np.sum(cash_flows / (1 + rate) ** (days / 21.0))
    try:
        return brentq(npv, 0, 1.0)
    except (RuntimeError, ValueError):
        return np.nan

print("\n" + "="*80)
print("INICIANDO C√ÅLCULO DA TAXA INTERNA DE RETORNO (TIR)")
print("="*80)

ref_date = df_report['DataGeracao'].max()
print(f"Data de Refer√™ncia para o c√°lculo da TIR: {ref_date.strftime('%d/%m/%Y')}")

try:
    df_feriados = pd.read_excel(caminho_feriados)
    # Limpa os feriados de valores nulos, 
    #  converte para o tipo correto
    holidays = pd.to_datetime(df_feriados['Data'], errors='coerce').dropna().dt.date.values
    print(f"Sucesso: {len(holidays)} feriados carregados.")
except Exception as e:
    print(f"[WGAP] N√£o foi poss√≠vel carregar feriados: {e}. O c√°lculo de dias √∫teis pode ser impreciso.")
    holidays = []

df_avencer = df_report[df_report['DataVencimento'] > ref_date].copy()

if not df_avencer.empty:
    try:
        # obs: Convertendo explicitamente para os tipos do NumPy
        start_date_np = np.datetime64(ref_date.date(), 'D')
        end_dates_np = df_avencer['DataVencimento'].dt.date.values.astype('datetime64[D]')
        holidays_np = np.array(holidays, dtype='datetime64[D]')

        df_avencer.loc[:, '_DIAS_UTEIS_'] = np.busday_count(start_date_np, end_dates_np, holidays=holidays_np)
        
        df_avencer = df_avencer[df_avencer['_DIAS_UTEIS_'] > 0]
        print("C√°lculo de dias √∫teis realizado com sucesso.")
    except Exception as e:
        print(f"[WGAP] Falha ao calcular dias √∫teis: {e}")
        df_avencer.loc[:, '_DIAS_UTEIS_'] = np.nan
else:
    print("[AVISO] N√£o h√° parcelas futuras ('a vencer') na base de dados. O c√°lculo de TIR ser√° pulado.")
    df_avencer['_DIAS_UTEIS_'] = np.nan


all_tirs = []
cat_cols = list(dimensoes_analise.values())
segmentos_para_analise = [('Carteira Total', 'Todos')] + \
                          [(col, seg) for col in cat_cols if col in df_avencer.columns for seg in df_avencer[col].dropna().unique()]

for tipo_dimensao, segmento in segmentos_para_analise:
    if tipo_dimensao == 'Carteira Total':
        df_segmento = df_avencer.copy()
    else:
        df_segmento = df_avencer[df_avencer[tipo_dimensao] == segmento].copy()

    if df_segmento.empty or df_segmento['_DIAS_UTEIS_'].isnull().all():
        continue

    vp_bruto = df_segmento['ValorPresente'].sum()
    tir_bruta, tir_pdd = np.nan, np.nan

    if vp_bruto > 0:
        #? TIR Bruta
        fluxos_brutos = df_segmento.groupby('_DIAS_UTEIS_')['ValorNominal'].sum()
        tir_bruta = calculate_xirr([-vp_bruto] + fluxos_brutos.values.tolist(), [0] + fluxos_brutos.index.tolist())

        #? TIR L√≠quida de PDD
        pdd_total_segmento = df_segmento['PDDTotal'].sum()
        pdd_rate = (pdd_total_segmento / vp_bruto) if pd.notna(pdd_total_segmento) and vp_bruto > 0 else 0.0
        df_segmento['Fluxo_PDD'] = df_segmento['ValorNominal'] * (1 - pdd_rate)
        fluxos_pdd = df_segmento.groupby('_DIAS_UTEIS_')['Fluxo_PDD'].sum()
        tir_pdd = calculate_xirr([-vp_bruto] + fluxos_pdd.values.tolist(), [0] + fluxos_pdd.index.tolist())

    all_tirs.append({
        'DimensaoColuna': tipo_dimensao, 'Segmento': segmento,
        'Valor Presente TIR (M)': vp_bruto / 1e6,
        'TIR Bruta \n(% a.m. )': tir_bruta * 100 if pd.notna(tir_bruta) else np.nan,
        'TIR L√≠quida de PDD \n(% a.m. )': tir_pdd * 100 if pd.notna(tir_pdd) else np.nan
    })

df_tir_summary = pd.DataFrame(all_tirs)
if not df_tir_summary.empty:
    tir_cols_to_fill = [col for col in df_tir_summary.columns if 'TIR' in col]
    df_tir_summary[tir_cols_to_fill] = df_tir_summary[tir_cols_to_fill].fillna(-100.0)

print("C√°lculo de TIR conclu√≠do.")
```

```python
# -*- coding: utf-8 -*-
"""
## An√°lise da Carteira de receb√≠veis (Vers√£o Final)

Script completo para an√°lise de estoque, com todas as corre√ß√µes de l√≥gica
implementadas, incluindo o ajuste no filtro de data para o c√°lculo da TIR.
"""

#* ==============>>>>>>>>
#* Bibliotecas   >>>>>>>>
#* ==============>>>>>>>>
import pandas as pd
import numpy as np
import os
from scipy.optimize import brentq
import base64
from IPython.display import display

pd.options.display.max_columns = 100
pd.options.display.max_rows = 200

"""##### <span style="color:#CFFFE5;">Par√¢metro de Neg√≥cio</span>"""

DIAS_ATRASO_DEFINICAO_VENCIDO = 1

"""### <span style="color:#AEE5F9;"> Leitura e Prepara√ß√£o dos Dados
<span style="color: #FFB3B3; font-size: 15px; font-weight: bold;">
    ATEN√á√ÉO: REDIFINIR AQUI OS PATHS E NOMES DE ARQUIVOS, SE NECESS√ÅRIO.
</span>
"""

# =============================================================================
# LER DADOS   =================================================================
# =============================================================================

#! PATHS ----------------------------------------------------------------------
DIRETORIO_DADOS = r'C:\Users\Leo\Desktop\Porto_Real\portoauto\src\vortx_estoques\data\131752-Estoque'
ARQUIVOS_CSV = [
    'FIDC FCT CONSIGNADO II  - Estoque 05.09.25-Parte1.csv',
    'FIDC FCT CONSIGNADO II  - Estoque 05.09.25-Parte2.csv',
    'FIDC FCT CONSIGNADO II  - Estoque 05.09.25-Parte3.csv'
]
caminho_feriados = r'C:\Users\Leo\Desktop\Porto_Real\portoauto\feriados_nacionais.xls'
logo_path = r'C:\Users\Leo\Desktop\Porto_Real\portoauto\images\logo_inv.png'
output_path = r'C:\Users\Leo\Desktop\Porto_Real\portoauto\src\vortx_estoques\output'
#! ----------------------------------------------------------------------------

# --- Carga e unifica√ß√£o dos arquivos ---
lista_dfs = []
caminhos_completos = [os.path.join(DIRETORIO_DADOS, arq) for arq in ARQUIVOS_CSV]
for caminho in caminhos_completos:
    df_temp = pd.read_csv(caminho, sep=';', decimal=',', dayfirst=True, encoding='latin1')
    lista_dfs.append(df_temp)
df_merged = pd.concat(lista_dfs, ignore_index=True)
print(f"Dados unificados com sucesso. Total de {df_merged.shape[0]} registros carregados.")

# --- Limpeza e Prepara√ß√£o dos Dados ---
print("Iniciando limpeza e prepara√ß√£o dos dados...")
colunas_para_remover = [
    'NumeroBoleto', 'CampoAdicional1', 'CampoAdicional2', 'CampoAdicional3', 'CampoAdicional4',
    'CampoAdicional5', 'Registradora', 'IdContratoRegistradora', 'IdTituloRegistradora',
    'Coobricacao', 'PES_TIPO_PESSOA', 'TIT_CEDENTE_ENT_CODIGO', 'Cnae',
    'SecaoCNAEDescricao', 'NotaPdd', 'SAC_TIPO_PESSOA'
]
df_merged.drop(columns=colunas_para_remover, inplace=True, errors='ignore')

# Renomear colunas para compatibilidade
df_merged = df_merged.rename(columns={'IdTituloVortxOriginador': 'CCB', 'SacadoCnpjCpf': 'SacadoID'})

# Convers√£o de tipos de dados
cols_data = ['DataGeracao', 'DataAquisicao', 'DataVencimento', 'DataEmissao']
for col in cols_data:
    df_merged[col] = pd.to_datetime(df_merged[col], errors='coerce', dayfirst=True)

df_final2 = df_merged.copy()
del df_merged, lista_dfs
print(f"DataFrame pronto para an√°lise com {df_final2.shape[0]} linhas.")


"""### <span style="color:#AEE5F9;"> Colunas Auxiliares e Novas Segmenta√ß√µes

Cria√ß√£o de colunas derivadas para an√°lise de risco e novas segmenta√ß√µes de neg√≥cio.
"""

# =============================================================================
#  Colunas Auxiliares =======================================================
# =============================================================================

data_referencia_relatorio = df_final2['DataGeracao'].max()
print(f"Usando a data do relat√≥rio como refer√™ncia para inadimpl√™ncia: {data_referencia_relatorio.strftime('%d/%m/%Y')}")

df_final2['_ValorLiquido'] = df_final2['ValorPresente'] - df_final2['PDDTotal']

dias_de_atraso = (data_referencia_relatorio - df_final2['DataVencimento']).dt.days
df_final2['_ParcelaVencida_Flag'] = ((dias_de_atraso >= DIAS_ATRASO_DEFINICAO_VENCIDO) & (dias_de_atraso > 0)).astype(int)
contratos_com_parcela_vencida = df_final2.groupby('CCB')['_ParcelaVencida_Flag'].max()
lista_ccbs_vencidos = contratos_com_parcela_vencida[contratos_com_parcela_vencida == 1].index
df_final2['_ContratoVencido_Flag'] = df_final2['CCB'].isin(lista_ccbs_vencidos).astype(int)

df_final2['_FlagDesagio'] = np.where(df_final2['ValorAquisicao'] > df_final2['ValorNominal'], '√Ågio', 'Des√°gio')

print("Cria√ß√£o de colunas auxiliares conclu√≠da.")
df_report = df_final2.copy()
del df_final2

"""### <span style="color:#AEE5F9;"> Qualidade e Consist√™ncia dos Dados

Verifico inconsist√™ncias, dados faltantes e caracter√≠sticas gerais da carteira.
"""
# =============================================================================
# Qualidade e Consist√™ncia dos Dados ==========================================
# =============================================================================
print("\n" + "="*80)
print("INICIANDO VERIFICA√á√ïES DE SANIDADE E QUALIDADE DOS DADOS")
print("="*80)
checks_results = {}
valor_presente_formatado = f"R$ {df_report['ValorPresente'].sum():,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.')
total_registros_formatado = f"{len(df_report):,}".replace(',', '.')
clientes_unicos_formatado = f"{df_report['SacadoID'].nunique():,}".replace(',', '.')
ccbs_unicos_formatado = f"{df_report['CCB'].nunique():,}".replace(',', '.')
checks_results['N√∫mero Total de Registros'] = total_registros_formatado
checks_results['Valor Presente Total da Carteira'] = valor_presente_formatado
checks_results['Per√≠odo da Carteira (Data de Aquisi√ß√£o)'] = f"{df_report['DataAquisicao'].min().strftime('%d/%m/%Y')} a {df_report['DataAquisicao'].max().strftime('%d/%m/%Y')}"
checks_results['N√∫mero de Clientes √önicos'] = clientes_unicos_formatado
checks_results['N√∫mero de Contratos √önicos'] = ccbs_unicos_formatado
checks_results['Valores Monet√°rios Negativos'] = (df_report[['ValorAquisicao', 'ValorNominal', 'ValorPresente', 'PDDTotal']] < 0).any().any()
checks_results['VP > Valor Nominal'] = (df_report['ValorPresente'] > df_report['ValorNominal']).sum()
checks_results['Valor Aquisi√ß√£o > Valor Nominal'] = (df_report['ValorAquisicao'] > df_report['ValorNominal']).sum()
checks_results['Data Aquisi√ß√£o > Data Vencimento'] = (df_report['DataAquisicao'] > df_report['DataVencimento']).sum()
print("Verifica√ß√µes de sanidade conclu√≠das.")


"""### <span style="color:#AEE5F9;"> Gera√ß√£o do Relat√≥rio Final

Consolida√ß√£o de todas as an√°lises em um relat√≥rio HTML interativo.
"""
# =============================================================================
# Gera√ß√£o do Relat√≥rio  ======================================================
# =============================================================================
dimensoes_analise = {
    'Cedentes': 'CedenteNome',
    'Tipo de Ativo': 'TipoAtivo',
    'Situa√ß√£o': 'Situacao',
    'Pagamento Parcial': 'PagamentoParcial',
    'Flag √Ågio/Des√°gio': '_FlagDesagio'
}
dimensoes_analise = {k: v for k, v in dimensoes_analise.items() if v in df_report.columns}
os.makedirs(output_path, exist_ok=True)

#***********************
#* C√ÅLCULO M√âTRICAS
#***********************
print("\n" + "="*80)
print("INICIANDO C√ÅLCULO UNIFICADO DAS M√âTRICAS")
print("="*80)

dias_de_atraso = (data_referencia_relatorio - df_report['DataVencimento']).dt.days
df_report['_ValorVencido_1d'] = df_report.loc[dias_de_atraso >= 1, 'ValorPresente'].fillna(0)
df_report['_ValorVencido_30d'] = df_report.loc[dias_de_atraso >= 30, 'ValorPresente'].fillna(0)
df_report['_ValorVencido_60d'] = df_report.loc[dias_de_atraso >= 60, 'ValorPresente'].fillna(0)
df_report.fillna({'_ValorVencido_1d': 0, '_ValorVencido_30d': 0, '_ValorVencido_60d': 0}, inplace=True)

vp_col_name = 'Valor Presente \n(R$ MM)'
vl_col_name = 'Valor L√≠quido \n(R$ MM)'
col_contratos_venc_perc = f"% Contratos Venc. (>{DIAS_ATRASO_DEFINICAO_VENCIDO}d)"
tabelas_metricas = {}

for nome_analise, coluna in dimensoes_analise.items():
    print(f"Calculando m√©tricas para a dimens√£o: '{nome_analise}'...")
    grouped = df_report.groupby(coluna, observed=True)
    total_contratos_unicos = grouped['CCB'].nunique()
    contratos_vencidos_unicos = df_report[df_report['_ContratoVencido_Flag'] == 1].groupby(coluna, observed=True)['CCB'].nunique()
    df_metricas = pd.DataFrame({'N¬∫ Contratos √önicos': total_contratos_unicos})
    somas_financeiras = grouped[['_ValorLiquido', 'ValorPresente', '_ValorVencido_1d', '_ValorVencido_30d', '_ValorVencido_60d']].sum()
    df_metricas = df_metricas.join(somas_financeiras)
    df_metricas['%PDD'] = (1 - df_metricas['_ValorLiquido'] / df_metricas['ValorPresente']) * 100
    df_metricas[col_contratos_venc_perc] = (contratos_vencidos_unicos.reindex(df_metricas.index).fillna(0) / df_metricas['N¬∫ Contratos √önicos']).fillna(0) * 100
    df_metricas['Venc. 1d / VP'] = (df_metricas['_ValorVencido_1d'] / df_metricas['ValorPresente']) * 100
    df_metricas['Venc. 30d / VP'] = (df_metricas['_ValorVencido_30d'] / df_metricas['ValorPresente']) * 100
    df_metricas['Venc. 60d / VP'] = (df_metricas['_ValorVencido_60d'] / df_metricas['ValorPresente']) * 100
    df_metricas = df_metricas.rename(columns={'ValorPresente': vp_col_name, '_ValorLiquido': vl_col_name, 'N¬∫ Contratos √önicos': 'N¬∫ Contratos'})
    df_metricas[[vp_col_name, vl_col_name]] /= 1e6
    df_metricas = df_metricas.drop(columns=['_ValorLiquido', '_ValorVencido_1d', '_ValorVencido_30d', '_ValorVencido_60d'], errors='ignore')
    tabelas_metricas[nome_analise] = df_metricas
print("C√°lculo unificado de m√©tricas conclu√≠do.")

#***********************
#* TICKET M√âDIO
#***********************
tabelas_ticket = {}
for nome_analise, coluna in dimensoes_analise.items():
    df_temp = df_report.dropna(subset=[coluna, 'ValorPresente', 'ValorNominal'])
    if df_temp.empty: continue
    grouped = df_temp.groupby(coluna, observed=True)
    numerador = grouped.apply(lambda g: (g['ValorNominal'] * g['ValorPresente']).sum(), include_groups=False)
    denominador = grouped['ValorPresente'].sum()
    ticket_ponderado = (numerador / denominador).replace([np.inf, -np.inf], 0)
    ticket_ponderado.name = "Ticket Ponderado (R$)"
    tabelas_ticket[nome_analise] = pd.DataFrame(ticket_ponderado)
print("Ticket m√©dio calculado.")

"""#### <span style="color:#FFDAC1;"> C√°lculo da TIR

C√°lculo da TIR (XIRR) para a carteira a vencer. Como n√£o h√° dados de custo, calculamos a TIR Bruta e a TIR L√≠quida de PDD.
"""
#******************
#* TIR com brentq
#*******************
def calculate_xirr(cash_flows, days):
    cash_flows = np.array(cash_flows)
    days = np.array(days)
    def npv(rate):
        if rate <= -1: return float('inf')
        with np.errstate(divide='ignore', over='ignore'):
            return np.sum(cash_flows / (1 + rate) ** (days / 21.0))
    try:
        return brentq(npv, 0, 1.0)
    except (RuntimeError, ValueError):
        return np.nan

print("\n" + "="*80)
print("INICIANDO C√ÅLCULO DA TAXA INTERNA DE RETORNO (TIR)")
print("="*80)

ref_date = df_report['DataGeracao'].max()
print(f"Data de Refer√™ncia para o c√°lculo da TIR: {ref_date.strftime('%d/%m/%Y')}")

try:
    df_feriados = pd.read_excel(caminho_feriados)
    holidays = pd.to_datetime(df_feriados['Data'], errors='coerce').dropna().dt.date.values
    print(f"Sucesso: {len(holidays)} feriados carregados.")
except Exception as e:
    print(f"[AVISO] N√£o foi poss√≠vel carregar feriados: {e}. O c√°lculo de dias √∫teis pode ser impreciso.")
    holidays = []

# CORRE√á√ÉO CR√çTICA: Alterado de '>' para '>=' para incluir parcelas que vencem no mesmo dia da refer√™ncia.
df_avencer = df_report[df_report['DataVencimento'] >= ref_date].copy()


if not df_avencer.empty:
    try:
        start_date_np = np.datetime64(ref_date.date(), 'D')
        end_dates_np = df_avencer['DataVencimento'].dt.date.values.astype('datetime64[D]')
        holidays_np = np.array(holidays, dtype='datetime64[D]')

        df_avencer.loc[:, '_DIAS_UTEIS_'] = np.busday_count(start_date_np, end_dates_np, holidays=holidays_np)
        
        # O filtro de dias √∫teis > 0 foi removido para incluir o fluxo no dia 0
        print("C√°lculo de dias √∫teis realizado com sucesso.")
    except Exception as e:
        print(f"[ERRO] Falha ao calcular dias √∫teis: {e}")
        df_avencer.loc[:, '_DIAS_UTEIS_'] = np.nan
else:
    print("[AVISO] N√£o h√° parcelas futuras ('a vencer') na base de dados. O c√°lculo de TIR ser√° pulado.")
    df_avencer['_DIAS_UTEIS_'] = np.nan


all_tirs = []
cat_cols = list(dimensoes_analise.values())
segmentos_para_analise = [('Carteira Total', 'Todos')] + \
                          [(col, seg) for col in cat_cols if col in df_avencer.columns for seg in df_avencer[col].dropna().unique()]

for tipo_dimensao, segmento in segmentos_para_analise:
    if tipo_dimensao == 'Carteira Total':
        df_segmento = df_avencer.copy()
    else:
        df_segmento = df_avencer[df_avencer[tipo_dimensao] == segmento].copy()

    if df_segmento.empty or df_segmento['_DIAS_UTEIS_'].isnull().all():
        continue

    vp_bruto = df_segmento['ValorPresente'].sum()
    tir_bruta, tir_pdd = np.nan, np.nan

    if vp_bruto > 0:
        fluxos_brutos = df_segmento.groupby('_DIAS_UTEIS_')['ValorNominal'].sum()
        tir_bruta = calculate_xirr([-vp_bruto] + fluxos_brutos.values.tolist(), [0] + fluxos_brutos.index.tolist())

        pdd_total_segmento = df_segmento['PDDTotal'].sum()
        pdd_rate = (pdd_total_segmento / vp_bruto) if pd.notna(pdd_total_segmento) and vp_bruto > 0 else 0.0
        df_segmento['Fluxo_PDD'] = df_segmento['ValorNominal'] * (1 - pdd_rate)
        fluxos_pdd = df_segmento.groupby('_DIAS_UTEIS_')['Fluxo_PDD'].sum()
        tir_pdd = calculate_xirr([-vp_bruto] + fluxos_pdd.values.tolist(), [0] + fluxos_pdd.index.tolist())

    all_tirs.append({
        'DimensaoColuna': tipo_dimensao, 'Segmento': segmento,
        'Valor Presente TIR (M)': vp_bruto / 1e6,
        'TIR Bruta \n(% a.m. )': tir_bruta * 100 if pd.notna(tir_bruta) else np.nan,
        'TIR L√≠quida de PDD \n(% a.m. )': tir_pdd * 100 if pd.notna(tir_pdd) else np.nan
    })

df_tir_summary = pd.DataFrame(all_tirs)
if not df_tir_summary.empty:
    tir_cols_to_fill = [col for col in df_tir_summary.columns if 'TIR' in col]
    df_tir_summary[tir_cols_to_fill] = df_tir_summary[tir_cols_to_fill].fillna(-100.0)

print("C√°lculo de TIR conclu√≠do.")

"""#### <span style="color:#FFDAC1;">  Montagem do HTML

Uno todos os elementos em um relat√≥rio HTML interativo.
"""
#***********************
#* GERA√á√ÉO DO RELAT√ìRIO HTML
#***********************

print("\n" + "="*80)
print("GERANDO RELAT√ìRIO HTML FINAL")
print("="*80)

def encode_image_to_base64(image_path):
    try:
        with open(image_path, "rb") as image_file:
            return base64.b64encode(image_file.read()).decode('utf-8')
    except FileNotFoundError:
        print(f"[ATEN√á√ÉO] Arquivo de imagem n√£o encontrado em: {image_path}. A logo n√£o ser√° exibida.")
        return None

logo_base64 = encode_image_to_base64(logo_path)
report_date = ref_date.strftime('%d/%m/%Y')

html_css = """
<style>
    /* Estilos para a mensagem de entrada √∫nica */
    .single-entry-message { background-color: #eef7f7; border-left: 5px solid #76c6c5; padding: 15px 20px; margin: 10px 0; font-size: 1.0em; color: #313131; }
    .single-entry-message strong { color: #163f3f; }
    .summary-title { font-size: 1.4em; color: #163f3f; border-bottom: 2px solid #76c6c5; padding-bottom: 10px; margin-top: 20px; margin-bottom: 20px; }
    .checks-container { display: grid; grid-template-columns: repeat(auto-fit, minmax(350px, 1fr)); gap: 15px; margin-bottom: 20px; }
    .check-item { background-color: #f5f5f5; padding: 12px; border-radius: 5px; border-left: 4px solid #76c6c5; font-size: 0.95em; }
    .check-item strong { color: #163f3f; }
    .dataframe-checks th { background-color: #5b8c8c; }
    .check-table-wrapper h4 { margin-top: 20px; margin-bottom: 10px; color: #163f3f; border-bottom: 2px solid #eeeeee; padding-bottom: 5px; }
    body { font-family: "Gill Sans MT", Arial, sans-serif; background-color: #f9f9f9; color: #313131; margin: 0; padding: 0; }
    .main-content { padding: 25px; }
    header { background-color: #163f3f; color: #FFFFFF; padding: 20px 40px; display: flex; justify-content: space-between; align-items: center; border-bottom: 5px solid #76c6c5; }
    header .logo img { height: 75px; }
    header .report-title { text-align: left; font-family: "Gill Sans MT", Arial, sans-serif; }
    header .report-title h1, header .report-title h2, header .report-title h3 { margin: 0; padding: 0; font-weight: normal; }
    header .report-title h1 { font-size: 1.6em; }
    header .report-title h2 { font-size: 1.4em; color: #d0d0d0; }
    header .report-title h3 { font-size: 1.1em; color: #a0a0a0; }
    .container-botoes { display: flex; flex-wrap: wrap; gap: 15px; margin-bottom: 25px; margin-top: 40px; }
    .container-botoes > details { flex: 1 1 280px; border: 1px solid #76c6c5; border-radius: 8px; overflow: hidden; }
    .container-botoes > details[open] { flex-basis: 100%; }
    details summary { font-size: 1.1em; font-weight: bold; color: #FFFFFF; background-color: #163f3f; padding: 15px 20px; cursor: pointer; outline: none; list-style-type: none; }
    details summary:hover { background-color: #0e5d5f; }
    details[open] summary { background-color: #76c6c5; color: #313131; }
    details[open] summary:hover { filter: brightness(95%); }
    summary::-webkit-details-marker { display: none; }
    summary::before { content: '‚ñ∫ '; margin-right: 8px; font-size: 0.8em;}
    details[open] summary::before { content: '‚ñº '; }
    .content-wrapper { padding: 20px; background-color: #FFFFFF; overflow-x: auto; border: 1px solid #ddd; border-radius: 8px; }
    table.dataframe, th, td { border: 1px solid #bbbbbb; }
    table.dataframe { border-collapse: collapse; width: 100%; }
    th, td { text-align: left; padding: 10px; vertical-align: middle; }
    th { background-color: #163f3f; color: #FFFFFF; }
    tr:nth-child(even) { background-color: #eeeeee; }
    footer { background-color: #f0f0f0; color: #555555; font-size: 0.8em; line-height: 1.6; padding: 25px 40px; margin-top: 40px; border-top: 1px solid #dddddd; }
    footer .disclaimer { margin-top: 20px; font-style: italic; border-top: 1px solid #dddddd; padding-top: 20px; }
</style>
"""

html_parts = ["<!DOCTYPE html><html lang='pt-BR'><head>",
              "<meta charset='UTF-8'><title>An√°lise de Desempenho - FCT Consignado II</title>",
              html_css, "</head><body>"]

html_parts.append(f"<header><div class='report-title'><h1>An√°lise de desempenho</h1><h2>FCT CONSIGNADO II</h2><h3>{report_date}</h3></div>")
if logo_base64:
    html_parts.append(f'<div class="logo"><img src="data:image/png;base64,{logo_base64}" alt="Logo"></div>')
html_parts.append("</header><div class='main-content'>")

mapa_descricoes = {
    'Cedentes': 'Analisa as m√©tricas de risco e retorno agrupadas por cada Cedente.',
    'Tipo de Ativo': 'Compara o desempenho entre os diferentes tipos de ativos (CCB, CT - Contrato).',
    'Situa√ß√£o': 'Compara o desempenho dos t√≠tulos com base na sua situa√ß√£o atual (Ex: Aditado).',
    'Pagamento Parcial': 'Verifica se h√° impacto nas m√©tricas para t√≠tulos que aceitam pagamento parcial.',
    'Flag √Ågio/Des√°gio': 'Compara o desempenho de contratos adquiridos com √°gio vs. des√°gio.'
}

# Verifica√ß√µes de Sanidade dos Dados
html_parts.append("<details open><summary>Verifica√ß√µes e Sanidade dos Dados</summary><div class='content-wrapper'>")
simple_checks_html = "<div class='checks-container'>"
table_checks_html = "<div class='check-table-wrapper'>"
for key, value in checks_results.items():
    if '[TABELA]' in key:
        clean_key = key.replace('[TABELA]', '').strip()
        table_checks_html += f"<h4>{clean_key}</h4>{str(value)}"
    else:
        simple_checks_html += f"<div class='check-item'><strong>{key}:</strong> {value}</div>"
simple_checks_html += "</div>"
table_checks_html += "</div>"
html_parts.append(simple_checks_html)
html_parts.append(table_checks_html)
html_parts.append("</div></details>")

# Resumo da Carteira Total
print("--> Gerando tabela de resumo da carteira total...")
total_metrics = {
    'N¬∫ Contratos': df_report['CCB'].nunique(),
    col_contratos_venc_perc: (df_report[df_report['_ContratoVencido_Flag'] == 1]['CCB'].nunique() / df_report['CCB'].nunique()) * 100,
    vl_col_name: df_report['_ValorLiquido'].sum() / 1e6,
    vp_col_name: df_report['ValorPresente'].sum() / 1e6,
    '%PDD': (1 - df_report['_ValorLiquido'].sum() / df_report['ValorPresente'].sum()) * 100 if df_report['ValorPresente'].sum() > 0 else 0,
    'Venc. 1d / VP': (df_report['_ValorVencido_1d'].sum() / df_report['ValorPresente'].sum()) * 100 if df_report['ValorPresente'].sum() > 0 else 0,
    'Venc. 30d / VP': (df_report['_ValorVencido_30d'].sum() / df_report['ValorPresente'].sum()) * 100 if df_report['ValorPresente'].sum() > 0 else 0,
    'Venc. 60d / VP': (df_report['_ValorVencido_60d'].sum() / df_report['ValorPresente'].sum()) * 100 if df_report['ValorPresente'].sum() > 0 else 0,
    'Ticket Ponderado (R$)': (df_report['ValorNominal'] * df_report['ValorPresente']).sum() / df_report['ValorPresente'].sum() if df_report['ValorPresente'].sum() > 0 else 0,
}
df_total = pd.DataFrame([total_metrics], index=['Total'])
if not df_tir_summary.empty:
    df_tir_total = df_tir_summary[df_tir_summary['DimensaoColuna'] == 'Carteira Total'].drop(columns=['DimensaoColuna', 'Segmento', 'Valor Presente TIR (M)'])
    if not df_tir_total.empty:
        df_tir_total.index = ['Total']
        df_total = pd.concat([df_total, df_tir_total], axis=1)


colunas_vencimento = ['Venc. 1d / VP', 'Venc. 30d / VP', 'Venc. 60d / VP']
colunas_ordem_total = ['N¬∫ Contratos', col_contratos_venc_perc, vl_col_name, vp_col_name, '%PDD'] + colunas_vencimento + ['Ticket Ponderado (R$)']
ordem_ideal_tir = ['TIR Bruta \n(% a.m. )', 'TIR L√≠quida de PDD \n(% a.m. )']
colunas_tir_ordenadas = [col for col in ordem_ideal_tir if col in df_total.columns]
colunas_finais_total = [col for col in colunas_ordem_total + colunas_tir_ordenadas if col in df_total.columns]
df_total = df_total[colunas_finais_total]

formatters = {
    vl_col_name: '{:,.2f}'.format, vp_col_name: '{:,.2f}'.format,
    'N¬∫ Contratos': lambda x: f'{x:,.0f}'.replace(',', '.'),
    'Ticket Ponderado (R$)': lambda x: f'R$ {x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'),
    '%PDD': '{:,.2f}%'.format
}
formatters[col_contratos_venc_perc] = '{:,.2f}%'.format
for col in colunas_vencimento: formatters[col] = '{:,.2f}%'.format
for col in colunas_tir_ordenadas: formatters[col] = '{:,.2f}%'.format

df_total.columns = [col.replace('\n', '<br>') for col in df_total.columns]

html_parts.append("<h2 class='summary-title'>Resumo da Carteira Total</h2><div class='content-wrapper'>")
html_parts.append(df_total.to_html(index=False, classes='dataframe', formatters=formatters, na_rep='-', escape=False))
html_parts.append("</div>")

# In√≠cio dos bot√µes retr√°teis
html_parts.append("<div class='container-botoes'>")
for nome_analise, coluna in dimensoes_analise.items():
    if coluna not in df_report.columns or df_report[coluna].isnull().all(): continue
    print(f"--> Processando e gerando HTML para: '{nome_analise}'")

    df_final = tabelas_metricas.get(nome_analise)
    df_ticket = tabelas_ticket.get(nome_analise, pd.DataFrame())

    df_final = df_final.join(df_ticket, how='outer')
    if not df_tir_summary.empty:
        df_tir = df_tir_summary[df_tir_summary['DimensaoColuna'] == coluna].set_index('Segmento')
        if not df_tir.empty:
            df_final = df_final.join(df_tir.drop(columns=['DimensaoColuna', 'Valor Presente TIR (M)']), how='outer')

    df_final.index.name = nome_analise
    df_final.reset_index(inplace=True)

    colunas_ordem = [nome_analise, 'N¬∫ Contratos', col_contratos_venc_perc, vl_col_name, vp_col_name, '%PDD'] + colunas_vencimento
    if 'Ticket Ponderado (R$)' in df_final.columns: colunas_ordem.append('Ticket Ponderado (R$)')

    colunas_finais = [col for col in colunas_ordem + colunas_tir_ordenadas if col in df_final.columns]
    outras_colunas = [col for col in df_final.columns if col not in colunas_finais]
    df_final = df_final[colunas_finais + outras_colunas]

    df_final = df_final.sort_values(vp_col_name, ascending=False).reset_index(drop=True)
    df_final.columns = [col.replace('\n', '<br>') for col in df_final.columns]

    html_parts.append("<details>")
    descricao = mapa_descricoes.get(nome_analise, 'Descri√ß√£o n√£o dispon√≠vel.')
    html_parts.append(f'<summary title="{descricao}">{nome_analise}</summary>')
    html_parts.append("<div class='content-wrapper'>")
    html_table = df_final.to_html(index=False, classes='dataframe', formatters=formatters, na_rep='-', escape=False)
    html_parts.append(html_table)
    html_parts.append("</div></details>")

html_parts.append("</div></div>")

footer_main_text = "Este documento tem como objetivo apresentar uma an√°lise de desempenho do fundo FCT Consignado II (CNPJ 52.203.615/0001-19)..."
footer_disclaimer = "Disclaimer: Este relat√≥rio foi preparado pelo Porto Real Investimentos exclusivamente para fins informativos..."
html_parts.append(f"<footer><p>{footer_main_text.strip()}</p><div class='disclaimer'>{footer_disclaimer.strip()}</div></footer>")
html_parts.append("</body></html>")

final_html_content = "\n".join(html_parts)
html_output_filename = os.path.join(output_path, 'analise_carteira_estoque.html')
try:
    with open(html_output_filename, 'w', encoding='utf-8') as f:
        f.write(final_html_content)
    print("\n" + "="*80)
    print("AN√ÅLISE CONCLU√çDA COM SUCESSO!")
    print(f"O relat√≥rio HTML final foi salvo em: {html_output_filename}")
    print("="*80)
except Exception as e:
    print(f"\n[ERRO GRAVE] N√£o foi poss√≠vel salvar o arquivo HTML: {e}")

#del df_report, df_total
```

```python
# -*- coding: utf-8 -*-
# =============================================================================
# C√âLULA DE DEBUG - AN√ÅLISE PROFUNDA DA TIR
# =============================================================================
# OBJETIVO: Entender por que os segmentos '√Ågio' e 'Pagamento Parcial = SIM'
# resultam em TIR Bruta de 0% e TIR L√≠quida de -100%.

import pandas as pd
import numpy as np
from scipy.optimize import brentq
from IPython.display import display

# --- 1. PREPARA√á√ÉO DO AMBIENTE DE DEBUG ---
# Recriando as vari√°veis e fun√ß√µes essenciais do seu script principal
# para que esta c√©lula possa ser executada de forma independente.

if 'df_report' not in locals():
    print("ERRO: O DataFrame 'df_report' n√£o foi encontrado. Por favor, execute o script principal primeiro.")
else:
    # Recriando a fun√ß√£o de c√°lculo da TIR
    def calculate_xirr(cash_flows, days):
        cash_flows = np.array(cash_flows)
        days = np.array(days)
        def npv(rate):
            if rate <= -1: return float('inf')
            with np.errstate(divide='ignore', over='ignore'):
                return np.sum(cash_flows / (1 + rate) ** (days / 21.0)) # Assumindo 21 dias √∫teis/m√™s
        try:
            # Ampliando os limites da busca para garantir que a raiz seja encontrada se existir
            return brentq(npv, -0.99, 5.0)
        except (RuntimeError, ValueError):
            return np.nan

    # Recriando vari√°veis de refer√™ncia do seu script
    ref_date = df_report['DataGeracao'].max()
    
    # Tentando carregar feriados, mas prosseguindo se n√£o encontrar
    try:
        df_feriados = pd.read_excel(caminho_feriados)
        holidays = pd.to_datetime(df_feriados['Data'], errors='coerce').dropna().dt.date.values
    except Exception:
        print("[AVISO] Arquivo de feriados n√£o encontrado. O c√°lculo de dias √∫teis prosseguir√° sem ele.")
        holidays = []
    
    # Filtrando para o DataFrame 'a vencer', que √© a base para a TIR
    df_avencer = df_report[df_report['DataVencimento'] >= ref_date].copy()
    
    # Calculando os dias √∫teis
    start_date_np = np.datetime64(ref_date.date(), 'D')
    end_dates_np = df_avencer['DataVencimento'].dt.date.values.astype('datetime64[D]')
    holidays_np = np.array(holidays, dtype='datetime64[D]')
    df_avencer['_DIAS_UTEIS_'] = np.busday_count(start_date_np, end_dates_np, holidays=holidays_np)
    
    print(f"Depura√ß√£o iniciada com base na data de refer√™ncia: {ref_date.strftime('%d/%m/%Y')}")
    print("-" * 80)

    # --- 2. AN√ÅLISE DO SEGMENTO "√ÅGIO" ---
    print("\n>>> ANALISANDO O SEGMENTO: _FlagDesagio == '√Ågio'")
    segmento_agio = df_avencer[df_avencer['_FlagDesagio'] == '√Ågio'].copy()
    
    if not segmento_agio.empty:
        # --- TIR BRUTA (√ÅGIO) ---
        vp_bruto_agio = segmento_agio['ValorPresente'].sum()
        fluxos_brutos_agio = segmento_agio.groupby('_DIAS_UTEIS_')['ValorNominal'].sum()

        print("\n--- TIR Bruta (√Ågio) ---")
        print(f"  - Valor Presente (Investimento): R$ {vp_bruto_agio:,.2f}")
        print(f"  - Soma dos Valores Nominais Futuros (Retorno): R$ {fluxos_brutos_agio.sum():,.2f}")
        
        if np.isclose(vp_bruto_agio, fluxos_brutos_agio.sum()):
             print("  - [DIAGN√ìSTICO] O Valor Presente total √© IGUAL ao Valor Nominal futuro.")
             print("  - Isso significa que o retorno financeiro (ganho de capital) √© zero.")
             print("  - Matematicamente, a √∫nica taxa que zera um fluxo de caixa sem ganho √© 0%.")
        
        tir_bruta_agio = calculate_xirr([-vp_bruto_agio] + fluxos_brutos_agio.values.tolist(), [0] + fluxos_brutos_agio.index.tolist())
        print(f"  - RESULTADO: TIR Bruta Calculada: {tir_bruta_agio*100 if pd.notna(tir_bruta_agio) else 'NaN':.2f}%")

        # --- TIR L√çQUIDA (√ÅGIO) ---
        print("\n--- TIR L√≠quida de PDD (√Ågio) ---")
        pdd_total_agio = segmento_agio['PDDTotal'].sum()
        pdd_rate_agio = pdd_total_agio / vp_bruto_agio if vp_bruto_agio > 0 else 0
        
        # Corrigindo o c√°lculo para aplicar a taxa de PDD ao fluxo de caixa bruto
        fluxos_pdd_agio = fluxos_brutos_agio * (1 - pdd_rate_agio)

        print(f"  - PDD Total do Segmento: R$ {pdd_total_agio:,.2f}")
        print(f"  - Taxa de PDD (PDD/VP): {pdd_rate_agio:.2%}")
        print(f"  - Fluxos de Caixa L√≠quidos Futuros (VN * (1-pdd_rate)): R$ {fluxos_pdd_agio.sum():,.2f}")

        if fluxos_pdd_agio.sum() < vp_bruto_agio:
            print("  - [DIAGN√ìSTICO] O fluxo de caixa L√çQUIDO esperado √© MENOR que o investimento.")
            print("  - Isso representa uma perda de capital. A TIR, neste caso, seria negativa.")
            print("  - O seu script substitui a falha no c√°lculo por -100%, o que explica o valor no relat√≥rio.")

        tir_pdd_agio = calculate_xirr([-vp_bruto_agio] + fluxos_pdd_agio.values.tolist(), [0] + fluxos_pdd_agio.index.tolist())
        tir_final_pdd_agio = (tir_pdd_agio * 100) if pd.notna(tir_pdd_agio) else -100.0
        print(f"  - RESULTADO: TIR L√≠quida Final (ap√≥s fillna): {tir_final_pdd_agio:.2f}%")
    else:
        print("  - Nenhuma parcela 'a vencer' encontrada para o segmento '√Ågio'.")

    print("-" * 80)

    # --- 3. AN√ÅLISE DO SEGMENTO "PAGAMENTO PARCIAL = SIM" ---
    print("\n>>> ANALISANDO O SEGMENTO: PagamentoParcial == 'SIM'")
    segmento_sim = df_avencer[df_avencer['PagamentoParcial'] == 'SIM'].copy()

    if not segmento_sim.empty:
        # --- TIR BRUTA (SIM) ---
        vp_bruto_sim = segmento_sim['ValorPresente'].sum()
        fluxos_brutos_sim = segmento_sim.groupby('_DIAS_UTEIS_')['ValorNominal'].sum()

        print("\n--- TIR Bruta (Pagamento Parcial = SIM) ---")
        print(f"  - Valor Presente (Investimento): R$ {vp_bruto_sim:,.2f}")
        print(f"  - Soma dos Valores Nominais Futuros (Retorno): R$ {fluxos_brutos_sim.sum():,.2f}")

        if np.isclose(vp_bruto_sim, fluxos_brutos_sim.sum()):
             print("  - [DIAGN√ìSTICO] O Valor Presente total √© IGUAL ao Valor Nominal futuro.")
             print("  - A TIR Bruta esperada √© 0%.")

        tir_bruta_sim = calculate_xirr([-vp_bruto_sim] + fluxos_brutos_sim.values.tolist(), [0] + fluxos_brutos_sim.index.tolist())
        print(f"  - RESULTADO: TIR Bruta Calculada: {tir_bruta_sim*100 if pd.notna(tir_bruta_sim) else 'NaN':.2f}%")

        # --- TIR L√çQUIDA (SIM) ---
        print("\n--- TIR L√≠quida de PDD (Pagamento Parcial = SIM) ---")
        pdd_total_sim = segmento_sim['PDDTotal'].sum()
        pdd_rate_sim = pdd_total_sim / vp_bruto_sim if vp_bruto_sim > 0 else 0
        fluxos_pdd_sim = fluxos_brutos_sim * (1 - pdd_rate_sim)

        print(f"  - PDD Total do Segmento: R$ {pdd_total_sim:,.2f}")
        print(f"  - Taxa de PDD (PDD/VP): {pdd_rate_sim:.2%}")
        print(f"  - Fluxos de Caixa L√≠quidos Futuros: R$ {fluxos_pdd_sim.sum():,.2f}")
        
        if fluxos_pdd_sim.sum() < vp_bruto_sim:
            print("  - [DIAGN√ìSTICO] O fluxo de caixa L√çQUIDO esperado √© MENOR que o investimento.")
            print("  - A TIR L√≠quida ser√° negativa, e o script a substitui por -100%.")

        tir_pdd_sim = calculate_xirr([-vp_bruto_sim] + fluxos_pdd_sim.values.tolist(), [0] + fluxos_pdd_sim.index.tolist())
        tir_final_pdd_sim = (tir_pdd_sim*100) if pd.notna(tir_pdd_sim) else -100.0
        print(f"  - RESULTADO: TIR L√≠quida Final (ap√≥s fillna): {tir_final_pdd_sim:.2f}%")
    else:
        print("  - Nenhuma parcela 'a vencer' encontrada para o segmento 'Pagamento Parcial = SIM'.")

    print("-" * 80)
    print("\nAN√ÅLISE DE DEBUG CONCLU√çDA.")
```

```python
# ---- Fun√ß√£o robusta para XIRR (substituir a calculate_xirr original) ----
import math

def calculate_xirr_safe(cash_flows, days, rate_upper=1.0, eps=1e-12):
    """
    cash_flows: lista/np.array de fluxos (com NEGATIVO no tempo 0: -vp)
    days: lista/np.array com os mesmos comprimentos, representando 'dias √∫teis' (inteiros)
    rate_upper: limite superior da busca (mantive 1.0 como no original)
    Retorna: taxa (por unidade usada na f√≥rmula: aqui, corresponde √† 'por 21 dias' usada no npv)
             ou np.nan se n√£o houver raiz numericamente detect√°vel.
    Observa√ß√µes: o c√°lculo usa discount (1+rate)**(days/21.0) same as original.
    """
    cf = np.array(cash_flows, dtype=float)
    d = np.array(days, dtype=float)

    # fun√ß√£o NPV com mesma conven√ß√£o (dividir por (1+rate)**(days/21))
    def npv_at(rate):
        if rate <= -1:
            return np.inf
        with np.errstate(divide='ignore', over='ignore', invalid='ignore'):
            return np.sum(cf / (1.0 + rate) ** (d / 21.0))

    # Checagem r√°pida: se todos os fluxos futuros s√£o zero -> sem raiz √∫til
    if np.allclose(cf[1:], 0.0, atol=eps):
        return np.nan

    # Avalia NPV em 0 (taxa zero)
    npv0 = npv_at(0.0)

    # Se npv0 == 0 (dentro de toler√¢ncia), a TIR √© 0.
    if np.isfinite(npv0) and abs(npv0) <= 1e-12:
        return 0.0

    # Avalia em limite superior
    try:
        npv_upper = npv_at(rate_upper)
    except Exception:
        npv_upper = np.nan

    # Se qualquer n√£o finito, devolve NaN
    if not np.isfinite(npv0) or not np.isfinite(npv_upper):
        # tenta brentq com um intervalo maior (como fallback), mas em geral devolvemos NaN
        try:
            return brentq(lambda r: npv_at(r), 1e-12, rate_upper)
        except Exception:
            return np.nan

    # Se sinais iguais, n√£o h√° garantia de raiz no intervalo (0, rate_upper)
    if npv0 * npv_upper > 0:
        # n√£o h√° mudan√ßa de sinal ‚Üí geralmente sem raiz √∫nica no intervalo
        # Se npv0 > 0 e npv_upper > 0 -> npv sempre positiva no intervalo (ex: VP < soma fluxos descontados)
        # se npv0 < 0 e npv_upper < 0 -> idem
        return np.nan

    # Caso tenha mudan√ßa de sinal, usa brentq com um limite m√≠nimo > 0 para evitar raiz em exatamente 0
    try:
        return brentq(lambda r: npv_at(r), 1e-12, rate_upper)
    except Exception:
        # tentativa final: se npv0 est√° muito pr√≥ximo de 0, retorna 0
        if abs(npv0) <= 1e-9:
            return 0.0
        return np.nan
```

```python
# ----- C√âLULA DE DEPURA√á√ÉO DA TIR (cole aqui ap√≥s df_avencer, ref_date e calculate_xirr estarem definidos) -----
import json

# 1) Prepara um dataframe de debug com as colunas relevantes
_df = df_avencer.copy()
cols_rel = [
    'CCB', 'SacadoID', 'PagamentoParcial', '_FlagDesagio', 'ValorPresente',
    'ValorNominal', 'PDDTotal', 'PDDNota', 'DataVencimento', 'DataGeracao', '_DIAS_UTEIS_'
]
for c in cols_rel:
    if c not in _df.columns:
        _df[c] = np.nan

# Normaliza strings para compara√ß√µes
_df['PagamentoParcial_up'] = _df['PagamentoParcial'].astype(str).str.strip().str.upper()
_df['_FlagDesagio_up'] = _df['_FlagDesagio'].astype(str).str.strip().str.upper()

# 2) Identifica casos suspeitos (v√°rios crit√©rios)
_cond_datas_iguais = _df['DataVencimento'].dt.date == ref_date.date()            # vencimento == refer√™ncia
_cond_dias_zero = _df['_DIAS_UTEIS_'] == 0                                     # dias √∫teis == 0
_cond_vp_eq_nom = np.isclose(_df['ValorPresente'].fillna(0), _df['ValorNominal'].fillna(0))  # VP == Nominal
_cond_vp_zero = _df['ValorPresente'] == 0
_cond_nom_zero = _df['ValorNominal'] == 0

_df['suspeito_datas_iguais'] = _cond_datas_iguais
_df['suspeito_dias_zero'] = _cond_dias_zero
_df['suspeito_vp_eq_nom'] = _cond_vp_eq_nom
_df['suspeito_vp_zero'] = _cond_vp_zero
_df['suspeito_nom_zero'] = _cond_nom_zero

# 3) Foca nos casos que o usu√°rio mencionou:
filtro_usuario = (_df['PagamentoParcial_up'] == 'SIM') & (_df['_FlagDesagio_up'] == '√ÅGIO'.upper())
# caso n√£o exista nenhum com exatamente 'SIM'/'√Ågio' (por varia√ß√µes), inclu√≠mos tamb√©m os suspeitos gerais
if _df[filtro_usuario].empty:
    filtro_usuario = (_df['suspeito_datas_iguais'] | _df['suspeito_dias_zero'] | _df['suspeito_vp_eq_nom'])

df_debug = _df[filtro_usuario].copy()

# 4) Agrupa por segmento (contrato) e recalcula fluxos exatamente como no c√°lculo original,
#    por√©m mantendo todos os detalhes para inspe√ß√£o
debug_rows = []
grouped = df_debug.groupby('CCB', observed=True)
for ccb, g in grouped:
    vp_bruto = g['ValorPresente'].sum()
    pdd_total_segmento = g['PDDTotal'].sum()
    pdd_rate = (pdd_total_segmento / vp_bruto) if (pd.notna(pdd_total_segmento) and vp_bruto > 0) else 0.0

    # fluxos brutos por dia √∫til
    fluxos_brutos = g.groupby('_DIAS_UTEIS_')['ValorNominal'].sum().sort_index()
    fluxos_pdd = (g.assign(Fluxo_PDD = g['ValorNominal'] * (1 - pdd_rate))
                    .groupby('_DIAS_UTEIS_')['Fluxo_PDD'].sum().sort_index())

    # marca se existe fluxo no dia 0 (al√©m do -vp inicial)
    fluxo_0_presente = (0 in fluxos_brutos.index and fluxos_brutos.loc[0] != 0)
    soma_fluxos_brutos = fluxos_brutos.sum()
    soma_fluxos_pdd = fluxos_pdd.sum()

    # tenta recalcular tir (sem captura de exce√ß√µes detalhadas)
    tir_bruta_raw = calculate_xirr([-vp_bruto] + fluxos_brutos.values.tolist(), [0] + fluxos_brutos.index.tolist()) if vp_bruto > 0 else np.nan
    tir_pdd_raw = calculate_xirr([-vp_bruto] + fluxos_pdd.values.tolist(), [0] + fluxos_pdd.index.tolist()) if vp_bruto > 0 else np.nan

    debug_rows.append({
        'CCB': ccb,
        'n_titulos_segmento': len(g),
        'vp_bruto': vp_bruto,
        'pdd_total_segmento': pdd_total_segmento,
        'pdd_rate': pdd_rate,
        'soma_fluxos_brutos': soma_fluxos_brutos,
        'soma_fluxos_pdd': soma_fluxos_pdd,
        'fluxo_0_presente': fluxo_0_presente,
        'fluxos_brutos_index_list': fluxos_brutos.index.tolist(),
        'fluxos_brutos_values_list': fluxos_brutos.values.tolist(),
        'fluxos_pdd_index_list': fluxos_pdd.index.tolist(),
        'fluxos_pdd_values_list': fluxos_pdd.values.tolist(),
        'tir_bruta_raw': tir_bruta_raw,
        'tir_pdd_raw': tir_pdd_raw
    })

df_debug_ccb = pd.DataFrame(debug_rows)

# 5) Pontos √≥bvios de alerta (crit√©rios compostos) ‚Äî cria colunas de flags
df_debug_ccb['flag_fluxo0'] = df_debug_ccb['fluxo_0_presente']
df_debug_ccb['flag_vp_eq_soma_fluxos'] = np.isclose(df_debug_ccb['vp_bruto'].fillna(0), df_debug_ccb['soma_fluxos_brutos'].fillna(0))
df_debug_ccb['flag_tir_numerica_zero'] = df_debug_ccb['tir_bruta_raw'].fillna(0) == 0
df_debug_ccb['flag_tir_nan_or_inf'] = df_debug_ccb['tir_bruta_raw'].isnull()

# Ordena por prioridade de suspeita
df_debug_ccb = df_debug_ccb.sort_values(
    by=['flag_fluxo0', 'flag_vp_eq_soma_fluxos', 'flag_tir_nan_or_inf', 'flag_tir_numerica_zero'],
    ascending=[False, False, False, False]
).reset_index(drop=True)

# 6) Exporta resultados detalhados para CSV para inspe√ß√£o local e imprime sum√°rio
out_debug_csv = os.path.join(output_path, 'debug_tir_ccb_details.csv')
df_debug_ccb.to_csv(out_debug_csv, index=False, encoding='utf-8-sig', sep=';')

print("Depura√ß√£o de TIR conclu√≠da. Resumo dos casos detectados:")
print(df_debug_ccb[['CCB', 'n_titulos_segmento', 'vp_bruto', 'pdd_total_segmento', 'pdd_rate',
                    'soma_fluxos_brutos', 'fluxo_0_presente', 'flag_vp_eq_soma_fluxos',
                    'tir_bruta_raw', 'tir_pdd_raw', 'flag_tir_nan_or_inf', 'flag_tir_numerica_zero']].head(30).to_string(index=False))

print(f"\nArquivo detalhado salvo em: {out_debug_csv}")
print("\nSugest√µes r√°pidas de verifica√ß√£o (manual):")
print("- Verifique t√≠tulos com 'fluxo_0_presente == True' (h√° pagamento nominal no dia de refer√™ncia).")
print("- Verifique contratos em que 'vp_bruto' ‚âà soma dos fluxos brutos (isso tende a gerar TIR pr√≥xima de 0).")
print("- Verifique registros com 'DataVencimento == DataGeracao' ou '_DIAS_UTEIS_ == 0' ‚Äî esses causam fluxos no tempo 0.")
print("- Verifique t√≠tulos com ValorPresente == ValorNominal e/ou PDDTotal == 0 (podem anular efeito de PDD no fluxo).")
print("\nSe quiser, rode esta c√©lula e me diga um CCB espec√≠fico listado no CSV para que eu gere um dump completo (linhas originais) daquele contrato.")
```

```python
# -*- coding: utf-8 -*-

"""
DEPURA√á√ÉO DA TIR - CASOS ESPEC√çFICOS
====================================

An√°lise detalhada dos casos onde:
1. Pagamento Parcial == "Sim" -> TIR Bruta = 0%
2. Flag √Ågio/Des√°gio == "√Ågio" -> TIR L√≠quida = -100%

"""

import pandas as pd
import numpy as np
from scipy.optimize import brentq

# Assumindo que df_report j√° existe do script principal
# Se n√£o existir, voc√™ precisar√° executar o script principal antes

print("="*80)
print("DEPURA√á√ÉO DA TIR - AN√ÅLISE DE CASOS ESPEC√çFICOS")
print("="*80)

# Fun√ß√£o TIR (copiada do script principal)
def calculate_xirr(cash_flows, days):
    cash_flows = np.array(cash_flows)
    days = np.array(days)
    
    def npv(rate):
        if rate <= -1: 
            return float('inf')
        with np.errstate(divide='ignore', over='ignore'):
            return np.sum(cash_flows / (1 + rate) ** (days / 21.0))
    
    try:
        return brentq(npv, 0, 1.0)
    except (RuntimeError, ValueError):
        return np.nan

# Data de refer√™ncia
ref_date = df_report['DataGeracao'].max()
print(f"Data de Refer√™ncia: {ref_date.strftime('%d/%m/%Y')}")

# Filtro para parcelas a vencer (>= data de refer√™ncia)
df_avencer = df_report[df_report['DataVencimento'] >= ref_date].copy()

print(f"\nTotal de registros a vencer: {len(df_avencer):,}")
print(f"Total de registros na base: {len(df_report):,}")

# ============================================================================
# 1. AN√ÅLISE DO CASO "PAGAMENTO PARCIAL == SIM"
# ============================================================================

print("\n" + "="*60)
print("1. AN√ÅLISE: PAGAMENTO PARCIAL == SIM")
print("="*60)

# Filtrar apenas os casos de pagamento parcial = SIM
df_pag_parcial_sim = df_avencer[df_avencer['PagamentoParcial'] == 'SIM'].copy()

print(f"Registros com Pagamento Parcial = SIM (a vencer): {len(df_pag_parcial_sim)}")

if len(df_pag_parcial_sim) > 0:
    print("\nEstat√≠sticas descritivas - Pagamento Parcial = SIM:")
    print("-" * 50)
    
    # C√°lculo de dias √∫teis
    start_date_np = np.datetime64(ref_date.date(), 'D')
    end_dates_np = df_pag_parcial_sim['DataVencimento'].dt.date.values.astype('datetime64[D]')
    df_pag_parcial_sim['_DIAS_UTEIS_'] = np.busday_count(start_date_np, end_dates_np)
    
    # Estat√≠sticas b√°sicas
    print(f"Valor Presente Total: R$ {df_pag_parcial_sim['ValorPresente'].sum():,.2f}")
    print(f"Valor Nominal Total: R$ {df_pag_parcial_sim['ValorNominal'].sum():,.2f}")
    print(f"N√∫mero de contratos √∫nicos: {df_pag_parcial_sim['CCB'].nunique()}")
    
    # An√°lise de datas
    print(f"\nAn√°lise de Datas:")
    print(f"Data de Gera√ß√£o (ref): {ref_date.strftime('%d/%m/%Y')}")
    print(f"Menor Data de Vencimento: {df_pag_parcial_sim['DataVencimento'].min().strftime('%d/%m/%Y')}")
    print(f"Maior Data de Vencimento: {df_pag_parcial_sim['DataVencimento'].max().strftime('%d/%m/%Y')}")
    
    # Verificar quantos t√™m data de vencimento = data de gera√ß√£o
    mesmo_dia = df_pag_parcial_sim[df_pag_parcial_sim['DataVencimento'].dt.date == ref_date.date()]
    print(f"Registros com vencimento no mesmo dia da gera√ß√£o: {len(mesmo_dia)}")
    
    if len(mesmo_dia) > 0:
        print(f"Valor Presente (mesmo dia): R$ {mesmo_dia['ValorPresente'].sum():,.2f}")
        print(f"Valor Nominal (mesmo dia): R$ {mesmo_dia['ValorNominal'].sum():,.2f}")
    
    # An√°lise de dias √∫teis
    print(f"\nDistribui√ß√£o de Dias √öteis:")
    dias_uteis_dist = df_pag_parcial_sim['_DIAS_UTEIS_'].value_counts().sort_index()
    for dias, count in dias_uteis_dist.items():
        valor_presente_dias = df_pag_parcial_sim[df_pag_parcial_sim['_DIAS_UTEIS_'] == dias]['ValorPresente'].sum()
        valor_nominal_dias = df_pag_parcial_sim[df_pag_parcial_sim['_DIAS_UTEIS_'] == dias]['ValorNominal'].sum()
        print(f"  {dias} dias √∫teis: {count} registros | VP: R$ {valor_presente_dias:,.2f} | VN: R$ {valor_nominal_dias:,.2f}")
    
    # C√°lculo manual da TIR para este segmento
    print(f"\nC√°lculo da TIR - Pagamento Parcial = SIM:")
    print("-" * 40)
    
    vp_total = df_pag_parcial_sim['ValorPresente'].sum()
    
    # Agrupamento por dias √∫teis
    fluxos_agrupados = df_pag_parcial_sim.groupby('_DIAS_UTEIS_')['ValorNominal'].sum()
    
    print(f"Fluxo inicial (VP negativo): -R$ {vp_total:,.2f}")
    print("Fluxos positivos por per√≠odo:")
    for dias, valor in fluxos_agrupados.items():
        print(f"  Dia {dias}: +R$ {valor:,.2f}")
    
    # C√°lculo da TIR
    if vp_total > 0:
        cash_flows = [-vp_total] + fluxos_agrupados.values.tolist()
        days = [0] + fluxos_agrupados.index.tolist()
        
        tir_result = calculate_xirr(cash_flows, days)
        print(f"\nTIR Calculada: {tir_result * 100 if pd.notna(tir_result) else 'ERRO':.4f}%")
        
        # Verifica√ß√£o manual para casos com dia 0
        if 0 in fluxos_agrupados.index:
            print(f"\n‚ö†Ô∏è  ATEN√á√ÉO: H√° fluxos no dia 0 (mesmo dia)!")
            fluxo_dia_0 = fluxos_agrupados[0]
            print(f"Fluxo no dia 0: +R$ {fluxo_dia_0:,.2f}")
            print(f"VP inicial: -R$ {vp_total:,.2f}")
            print(f"Resultado l√≠quido no dia 0: R$ {fluxo_dia_0 - vp_total:,.2f}")
            
            if abs(fluxo_dia_0 - vp_total) < 0.01:  # Praticamente zero
                print(">>> CAUSA IDENTIFICADA: Fluxo no dia 0 anula o investimento inicial!")
                print(">>> Isso resulta em TIR = 0% (sem ganho nem perda)")

else:
    print("‚ö†Ô∏è  Nenhum registro encontrado com Pagamento Parcial = SIM na carteira a vencer.")

# ============================================================================
# 2. AN√ÅLISE DO CASO "FLAG √ÅGIO/DES√ÅGIO == √ÅGIO"
# ============================================================================

print("\n" + "="*60)
print("2. AN√ÅLISE: FLAG √ÅGIO/DES√ÅGIO == √ÅGIO")
print("="*60)

# Filtrar apenas os casos de √°gio
df_agio = df_avencer[df_avencer['_FlagDesagio'] == '√Ågio'].copy()

print(f"Registros com √Ågio (a vencer): {len(df_agio)}")

if len(df_agio) > 0:
    print("\nEstat√≠sticas descritivas - √Ågio:")
    print("-" * 30)
    
    # C√°lculo de dias √∫teis
    start_date_np = np.datetime64(ref_date.date(), 'D')
    end_dates_np = df_agio['DataVencimento'].dt.date.values.astype('datetime64[D]')
    df_agio['_DIAS_UTEIS_'] = np.busday_count(start_date_np, end_dates_np)
    
    # Estat√≠sticas b√°sicas
    print(f"Valor Presente Total: R$ {df_agio['ValorPresente'].sum():,.2f}")
    print(f"Valor Nominal Total: R$ {df_agio['ValorNominal'].sum():,.2f}")
    print(f"PDD Total: R$ {df_agio['PDDTotal'].sum():,.2f}")
    print(f"Valor Aquisi√ß√£o Total: R$ {df_agio['ValorAquisicao'].sum():,.2f}")
    
    # Verifica√ß√£o da l√≥gica do √°gio
    print(f"\nVerifica√ß√£o da l√≥gica de √Ågio:")
    agio_check = df_agio['ValorAquisicao'] > df_agio['ValorNominal']
    print(f"Registros onde ValorAquisicao > ValorNominal: {agio_check.sum()} de {len(df_agio)}")
    
    # An√°lise de PDD
    pdd_total = df_agio['PDDTotal'].sum()
    vp_total = df_agio['ValorPresente'].sum()
    pdd_rate = pdd_total / vp_total if vp_total > 0 else 0
    
    print(f"Taxa de PDD m√©dia: {pdd_rate * 100:.2f}%")
    
    # C√°lculo manual da TIR L√≠quida para este segmento
    print(f"\nC√°lculo da TIR L√≠quida de PDD - √Ågio:")
    print("-" * 40)
    
    if vp_total > 0:
        # Fluxos com desconto de PDD
        df_agio['Fluxo_PDD'] = df_agio['ValorNominal'] * (1 - pdd_rate)
        fluxos_pdd_agrupados = df_agio.groupby('_DIAS_UTEIS_')['Fluxo_PDD'].sum()
        
        print(f"Fluxo inicial (VP negativo): -R$ {vp_total:,.2f}")
        print(f"Taxa de PDD aplicada: {pdd_rate * 100:.2f}%")
        print("Fluxos l√≠quidos de PDD por per√≠odo:")
        
        for dias, valor in fluxos_pdd_agrupados.items():
            valor_bruto = df_agio[df_agio['_DIAS_UTEIS_'] == dias]['ValorNominal'].sum()
            print(f"  Dia {dias}: R$ {valor_bruto:,.2f} -> R$ {valor:,.2f} (ap√≥s PDD)")
        
        # C√°lculo da TIR L√≠quida
        cash_flows_pdd = [-vp_total] + fluxos_pdd_agrupados.values.tolist()
        days = [0] + fluxos_pdd_agrupados.index.tolist()
        
        tir_liquida = calculate_xirr(cash_flows_pdd, days)
        print(f"\nTIR L√≠quida Calculada: {tir_liquida * 100 if pd.notna(tir_liquida) else 'ERRO':.4f}%")
        
        # Verifica√ß√£o se o VPL √© muito negativo
        if pd.notna(tir_liquida) and tir_liquida < -0.99:
            print(f"\n‚ö†Ô∏è  ATEN√á√ÉO: TIR muito negativa (pr√≥xima de -100%)!")
            print("Poss√≠veis causas:")
            
            total_fluxo_futuro = fluxos_pdd_agrupados.sum()
            print(f"Total de fluxos futuros (l√≠quido): R$ {total_fluxo_futuro:,.2f}")
            print(f"Investimento inicial: R$ {vp_total:,.2f}")
            print(f"Diferen√ßa: R$ {total_fluxo_futuro - vp_total:,.2f}")
            
            if total_fluxo_futuro < vp_total:
                print(">>> CAUSA IDENTIFICADA: Fluxos futuros < Investimento inicial")
                print(">>> PDD muito alta torna o investimento invi√°vel (TIR negativa)")

else:
    print("‚ö†Ô∏è  Nenhum registro encontrado com √Ågio na carteira a vencer.")

# ============================================================================
# 3. AN√ÅLISE CRUZADA: √ÅGIO + PAGAMENTO PARCIAL
# ============================================================================

print("\n" + "="*60)
print("3. AN√ÅLISE CRUZADA: √ÅGIO + PAGAMENTO PARCIAL")
print("="*60)

df_cruzado = df_avencer[
    (df_avencer['_FlagDesagio'] == '√Ågio') & 
    (df_avencer['PagamentoParcial'] == 'SIM')
].copy()

print(f"Registros com √Ågio + Pagamento Parcial = SIM: {len(df_cruzado)}")

if len(df_cruzado) > 0:
    print(f"VP Total (cruzado): R$ {df_cruzado['ValorPresente'].sum():,.2f}")
    print(f"VN Total (cruzado): R$ {df_cruzado['ValorNominal'].sum():,.2f}")
    print(f"PDD Total (cruzado): R$ {df_cruzado['PDDTotal'].sum():,.2f}")

# ============================================================================
# 4. RECOMENDA√á√ïES E CORRE√á√ïES
# ============================================================================

print("\n" + "="*60)
print("4. RECOMENDA√á√ïES E POSS√çVEIS CORRE√á√ïES")
print("="*60)

print("""
DIAGN√ìSTICO IDENTIFICADO:

1. TIR = 0% (Pagamento Parcial = SIM):
   - Provavelmente h√° fluxos no dia 0 (mesmo dia da gera√ß√£o)
   - Fluxos imediatos anulam o investimento inicial
   - Resultado: sem ganho nem perda = TIR = 0%

2. TIR = -100% (√Ågio):
   - PDD muito alta reduz drasticamente os fluxos futuros
   - Fluxos futuros (l√≠quidos de PDD) < Investimento inicial
   - Resultado: investimento invi√°vel = TIR muito negativa

POSS√çVEIS CORRE√á√ïES:

1. Para fluxos no dia 0:
   - Considerar mover para D+1
   - Ou excluir da an√°lise de TIR
   - Ou tratar como opera√ß√£o instant√¢nea

2. Para PDD alta:
   - Verificar se a taxa de PDD est√° correta
   - Considerar PDD espec√≠fica por segmento
   - Revisar metodologia de c√°lculo de PDD

3. Filtros adicionais:
   - Excluir opera√ß√µes com prazo < X dias
   - Excluir opera√ß√µes com PDD > Y%
   - Separar an√°lises por tipo de opera√ß√£o
""")

print("\n" + "="*80)
print("DEPURA√á√ÉO CONCLU√çDA")
print("="*80)
```

```python
# -*- coding: utf-8 -*-

"""
CORRE√á√ïES PARA O C√ÅLCULO DE TIR - IMPLEMENTA√á√ÉO
================================================

Substitua a se√ß√£o de c√°lculo de TIR no seu script principal 
por este c√≥digo corrigido.
"""

# ============================================================================
# C√ÅLCULO DE TIR CORRIGIDO COM FILTROS
# ============================================================================

def calculate_xirr(cash_flows, days):
    cash_flows = np.array(cash_flows)
    days = np.array(days)
    
    def npv(rate):
        if rate <= -1: 
            return float('inf')
        with np.errstate(divide='ignore', over='ignore'):
            return np.sum(cash_flows / (1 + rate) ** (days / 21.0))
    
    try:
        return brentq(npv, 0, 1.0)
    except (RuntimeError, ValueError):
        return np.nan

print("\n" + "="*80)
print("INICIANDO C√ÅLCULO DA TAXA INTERNA DE RETORNO (TIR) - VERS√ÉO CORRIGIDA")
print("="*80)

ref_date = df_report['DataGeracao'].max()
print(f"Data de Refer√™ncia para o c√°lculo da TIR: {ref_date.strftime('%d/%m/%Y')}")

# Carregar feriados
try:
    df_feriados = pd.read_excel(caminho_feriados)
    holidays = pd.to_datetime(df_feriados['Data'], errors='coerce').dropna().dt.date.values
    print(f"Sucesso: {len(holidays)} feriados carregados.")
except Exception as e:
    print(f"[AVISO] N√£o foi poss√≠vel carregar feriados: {e}. O c√°lculo de dias √∫teis pode ser impreciso.")
    holidays = []

# ============================================================================
# FILTROS PARA MELHORIA DA QUALIDADE DOS DADOS
# ============================================================================

print("\nüîß APLICANDO FILTROS DE QUALIDADE:")
print("-" * 50)

# Filtro base: parcelas a vencer
df_avencer_base = df_report[df_report['DataVencimento'] >= ref_date].copy()
print(f"1Ô∏è‚É£ Parcelas a vencer: {len(df_avencer_base):,} registros")

# Filtro 1: Prazo m√≠nimo de 30 dias
df_avencer_base['_DIAS_ATE_VENCIMENTO'] = (df_avencer_base['DataVencimento'] - ref_date).dt.days
df_avencer_prazo = df_avencer_base[df_avencer_base['_DIAS_ATE_VENCIMENTO'] >= 30].copy()
print(f"2Ô∏è‚É£ Com prazo >= 30 dias: {len(df_avencer_prazo):,} registros ({len(df_avencer_base) - len(df_avencer_prazo):,} filtrados)")

# Filtro 2: Opera√ß√µes com rendimento positivo
df_avencer_rendimento = df_avencer_prazo[df_avencer_prazo['ValorNominal'] > df_avencer_prazo['ValorPresente']].copy()
print(f"3Ô∏è‚É£ Com rendimento > 0: {len(df_avencer_rendimento):,} registros ({len(df_avencer_prazo) - len(df_avencer_rendimento):,} filtrados)")

# Filtro 3: PDD razo√°vel (<= 50%)
df_avencer_rendimento['_TAXA_PDD'] = df_avencer_rendimento['PDDTotal'] / df_avencer_rendimento['ValorPresente']
df_avencer_pdd_temp = df_avencer_rendimento[df_avencer_rendimento['_TAXA_PDD'] <= 0.50].copy()
print(f"4Ô∏è‚É£ Com PDD <= 50%: {len(df_avencer_pdd_temp):,} registros ({len(df_avencer_rendimento) - len(df_avencer_pdd_temp):,} filtrados)")

# Usar dataset com todos os filtros para TIR
df_avencer_filtrado = df_avencer_pdd_temp.copy()  # Dataset final com todos os filtros aplicados

# Copiar as colunas corrigidas do df_report para o dataset filtrado
# Primeiro criar um dicion√°rio de mapeamento CCB -> Flag corrigida
map_flag_corrigido = df_report.set_index('CCB')['_FlagDesagio_Corrigido'].to_dict()

# Aplicar o mapeamento ao dataset filtrado
df_avencer_filtrado['_FlagDesagio_Corrigido'] = df_avencer_filtrado['CCB'].map(map_flag_corrigido).fillna('Des√°gio')

print(f"üéØ FINAL - Registros para TIR: {len(df_avencer_filtrado):,}")
print(f"   Valor Presente: R$ {df_avencer_filtrado['ValorPresente'].sum():,.2f}")
print(f"   Valor Nominal: R$ {df_avencer_filtrado['ValorNominal'].sum():,.2f}")
print(f"   Rendimento Esperado: R$ {df_avencer_filtrado['ValorNominal'].sum() - df_avencer_filtrado['ValorPresente'].sum():,.2f}")

# Verificar distribui√ß√£o da classifica√ß√£o corrigida no dataset filtrado
print(f"\nDistribui√ß√£o da classifica√ß√£o corrigida (dataset filtrado):")
flag_dist = df_avencer_filtrado['_FlagDesagio_Corrigido'].value_counts()
for categoria, count in flag_dist.items():
    print(f"  {categoria}: {count:,} registros")

# ============================================================================
# CORRE√á√ÉO DA CLASSIFICA√á√ÉO √ÅGIO/DES√ÅGIO
# ============================================================================

print("\nüîß CORRIGINDO CLASSIFICA√á√ÉO √ÅGIO/DES√ÅGIO:")
print("-" * 50)

# Classifica√ß√£o original (problem√°tica)
agio_original = (df_report['_FlagDesagio'] == '√Ågio').sum()
desagio_original = (df_report['_FlagDesagio'] == 'Des√°gio').sum()

print(f"Classifica√ß√£o ORIGINAL:")
print(f"  √Ågio: {agio_original:,} registros")
print(f"  Des√°gio: {desagio_original:,} registros")

# Nova classifica√ß√£o baseada em ValorAquisicao vs ValorNominal com valida√ß√£o
df_report['_FlagDesagio_Corrigido'] = 'Des√°gio'  # Default para receb√≠veis

# Identificar casos suspeitos (ValorAquisicao muito maior que ValorNominal)
suspeitos_mask = df_report['ValorAquisicao'] > (df_report['ValorNominal'] * 2)  # Mais que 2x
df_report.loc[suspeitos_mask, '_FlagDesagio_Corrigido'] = 'Dados_Suspeitos'

# Para casos normais, manter a classifica√ß√£o baseada na regra original, mas validada
normal_mask = df_report['ValorAquisicao'] <= (df_report['ValorNominal'] * 2)
agio_normal_mask = normal_mask & (df_report['ValorAquisicao'] > df_report['ValorNominal'])
df_report.loc[agio_normal_mask, '_FlagDesagio_Corrigido'] = '√Ågio'

# Estat√≠sticas da nova classifica√ß√£o
desagio_novo = (df_report['_FlagDesagio_Corrigido'] == 'Des√°gio').sum()
agio_novo = (df_report['_FlagDesagio_Corrigido'] == '√Ågio').sum()
suspeitos_novo = (df_report['_FlagDesagio_Corrigido'] == 'Dados_Suspeitos').sum()

print(f"\nClassifica√ß√£o CORRIGIDA:")
print(f"  Des√°gio: {desagio_novo:,} registros")
print(f"  √Ågio (validado): {agio_novo:,} registros") 
print(f"  Dados Suspeitos: {suspeitos_novo:,} registros")

# ============================================================================
# C√ÅLCULO DE TIR COM DADOS FILTRADOS
# ============================================================================

if not df_avencer_filtrado.empty:
    try:
        # Calcular dias √∫teis
        start_date_np = np.datetime64(ref_date.date(), 'D')
        end_dates_np = df_avencer_filtrado['DataVencimento'].dt.date.values.astype('datetime64[D]')
        holidays_np = np.array(holidays, dtype='datetime64[D]')
        
        df_avencer_filtrado['_DIAS_UTEIS_'] = np.busday_count(start_date_np, end_dates_np, holidays=holidays_np)
        
        print("\n‚úÖ C√°lculo de dias √∫teis realizado com sucesso.")
    except Exception as e:
        print(f"\n‚ùå Erro ao calcular dias √∫teis: {e}")
        df_avencer_filtrado['_DIAS_UTEIS_'] = np.nan
else:
    print("\n‚ö†Ô∏è N√£o h√° parcelas v√°lidas para c√°lculo de TIR ap√≥s filtros.")

# Lista para armazenar TIRs calculadas
all_tirs_corrigidas = []
cat_cols = list(dimensoes_analise.values())

# Adicionar nova dimens√£o corrigida
dimensoes_analise_corrigidas = dimensoes_analise.copy()
dimensoes_analise_corrigidas['Flag √Ågio/Des√°gio (Corrigida)'] = '_FlagDesagio_Corrigido'

# Verificar se a coluna existe antes de usar
colunas_disponiveis = df_avencer_filtrado.columns.tolist()
print(f"Colunas dispon√≠veis no dataset filtrado: {len(colunas_disponiveis)}")

# Segmentos para an√°lise - s√≥ incluir _FlagDesagio_Corrigido se existir
segmentos_para_analise = [('Carteira Total', 'Todos')] + \
                        [(col, seg) for col in cat_cols if col in df_avencer_filtrado.columns 
                         for seg in df_avencer_filtrado[col].dropna().unique()]

# Adicionar segmentos da classifica√ß√£o corrigida se a coluna existir
if '_FlagDesagio_Corrigido' in df_avencer_filtrado.columns:
    segmentos_corrigidos = [('_FlagDesagio_Corrigido', seg) 
                           for seg in df_avencer_filtrado['_FlagDesagio_Corrigido'].dropna().unique()]
    segmentos_para_analise.extend(segmentos_corrigidos)
    print(f"Incluindo {len(segmentos_corrigidos)} segmentos da classifica√ß√£o corrigida")
else:
    print("Aviso: Coluna _FlagDesagio_Corrigido n√£o encontrada no dataset filtrado")

print(f"\nüéØ CALCULANDO TIR PARA {len(segmentos_para_analise)} SEGMENTOS:")
print("-" * 60)

for tipo_dimensao, segmento in segmentos_para_analise:
    if tipo_dimensao == 'Carteira Total':
        df_segmento = df_avencer_filtrado.copy()
        nome_segmento = 'Carteira Total'
    else:
        df_segmento = df_avencer_filtrado[df_avencer_filtrado[tipo_dimensao] == segmento].copy()
        nome_segmento = f"{tipo_dimensao}: {segmento}"
    
    if df_segmento.empty or df_segmento['_DIAS_UTEIS_'].isnull().all():
        continue
    
    vp_bruto = df_segmento['ValorPresente'].sum()
    tir_bruta, tir_pdd = np.nan, np.nan
    
    if vp_bruto > 10000:  # Filtro m√≠nimo de valor para TIR confi√°vel
        # TIR Bruta
        fluxos_brutos = df_segmento.groupby('_DIAS_UTEIS_')['ValorNominal'].sum()
        if len(fluxos_brutos) > 0:
            tir_bruta = calculate_xirr([-vp_bruto] + fluxos_brutos.values.tolist(), 
                                     [0] + fluxos_brutos.index.tolist())
        
        # TIR L√≠quida de PDD
        pdd_total_segmento = df_segmento['PDDTotal'].sum()
        pdd_rate = (pdd_total_segmento / vp_bruto) if pd.notna(pdd_total_segmento) and vp_bruto > 0 else 0.0
        
        if pdd_rate < 0.9:  # S√≥ calcular se PDD < 90%
            df_segmento['Fluxo_PDD'] = df_segmento['ValorNominal'] * (1 - pdd_rate)
            fluxos_pdd = df_segmento.groupby('_DIAS_UTEIS_')['Fluxo_PDD'].sum()
            if len(fluxos_pdd) > 0:
                tir_pdd = calculate_xirr([-vp_bruto] + fluxos_pdd.values.tolist(), 
                                       [0] + fluxos_pdd.index.tolist())
    
    all_tirs_corrigidas.append({
        'DimensaoColuna': tipo_dimensao,
        'Segmento': segmento,
        'Valor Presente TIR (M)': vp_bruto / 1e6,
        'TIR Bruta \n(% a.m. )': tir_bruta * 100 if pd.notna(tir_bruta) else np.nan,
        'TIR L√≠quida de PDD \n(% a.m. )': tir_pdd * 100 if pd.notna(tir_pdd) else np.nan
    })
    
    # Log do progresso para segmentos principais
    if tipo_dimensao in ['Carteira Total', 'CedenteNome'] or 'Corrigido' in tipo_dimensao:
        tir_bruta_str = f"{tir_bruta * 100:.2f}%" if pd.notna(tir_bruta) else "N/A"
        tir_pdd_str = f"{tir_pdd * 100:.2f}%" if pd.notna(tir_pdd) else "N/A"
        print(f"  ‚úÖ {nome_segmento[:50]:<50} | TIR: {tir_bruta_str} | TIR PDD: {tir_pdd_str}")

# Criar DataFrame final
df_tir_summary_corrigida = pd.DataFrame(all_tirs_corrigidas)

if not df_tir_summary_corrigida.empty:
    tir_cols_to_fill = [col for col in df_tir_summary_corrigida.columns if 'TIR' in col]
    df_tir_summary_corrigida[tir_cols_to_fill] = df_tir_summary_corrigida[tir_cols_to_fill].fillna(-100.0)
    
    # Mostrar resultados principais
    print(f"\nüéØ RESULTADOS PRINCIPAIS:")
    print("-" * 30)
    
    carteira_total = df_tir_summary_corrigida[df_tir_summary_corrigida['DimensaoColuna'] == 'Carteira Total']
    if not carteira_total.empty:
        tir_bruta_total = carteira_total['TIR Bruta \n(% a.m. )'].iloc[0]
        tir_pdd_total = carteira_total['TIR L√≠quida de PDD \n(% a.m. )'].iloc[0]
        
        if tir_bruta_total > -100:
            print(f"üèÜ TIR Bruta Carteira Total: {tir_bruta_total:.2f}% a.m. ({((1 + tir_bruta_total/100)**12 - 1)*100:.2f}% a.a.)")
        if tir_pdd_total > -100:
            print(f"üèÜ TIR L√≠quida Carteira Total: {tir_pdd_total:.2f}% a.m. ({((1 + tir_pdd_total/100)**12 - 1)*100:.2f}% a.a.)")

print(f"\n‚úÖ C√°lculo de TIR corrigido conclu√≠do.")
print(f"üìä {len(df_tir_summary_corrigida)} segmentos analisados.")

# Para usar no restante do script, substitua df_tir_summary por df_tir_summary_corrigida
# e adicione a dimens√£o corrigida √†s an√°lises
```

```python
# =============================================================================
# C√ìDIGO DE DEPURA√á√ÉO FINAL: PROVA DA TEORIA DE DUPLO EVENTO
# =============================================================================
print("\n" + "#"*80)
print("### DEPURA√á√ÉO FINAL: PROVANDO A TEORIA DOS DOIS TIPOS DE LIQUIDA√á√ÉO ###")
print("#"*80)

# 1. Isolar o universo de todas as parcelas que "terminam" na data do relat√≥rio
data_referencia_relatorio = df_report['DataGeracao'].max()
df_termino_hoje = df_report[df_report['DataVencimento'] == data_referencia_relatorio].copy()

print(f"\nAnalisando todas as {len(df_termino_hoje)} parcelas que vencem em {data_referencia_relatorio.strftime('%d/%m/%Y')}...")

# 2. Dividir este universo em dois grupos, baseando-se na teoria
df_saudavel = df_termino_hoje[df_termino_hoje['ValorPresente'] == df_termino_hoje['ValorNominal']].copy()
df_problematico = df_termino_hoje[df_termino_hoje['ValorPresente'] < df_termino_hoje['ValorNominal']].copy()

# 3. Fun√ß√£o para analisar e resumir cada grupo
def analisar_grupo(df, nome_grupo):
    if df.empty:
        return {
            'Grupo': nome_grupo,
            'N¬∫ Contratos √önicos': 0,
            'Valor Presente (R$)': 0,
            'PDD Total (R$)': 0,
            '% PDD': 0
        }
    
    num_contratos = df['CCB'].nunique()
    vp_total = df['ValorPresente'].sum()
    pdd_total = df['PDDTotal'].sum()
    pdd_rate = (pdd_total / vp_total * 100) if vp_total > 0 else 0
    
    return {
        'Grupo': nome_grupo,
        'N¬∫ Contratos √önicos': num_contratos,
        'Valor Presente (R$)': vp_total,
        'PDD Total (R$)': pdd_total,
        '% PDD': pdd_rate
    }

# 4. Analisar cada grupo e juntar os resultados
resultado_saudavel = analisar_grupo(df_saudavel, "Liquida√ß√£o Saud√°vel (VP=VN)")
resultado_problematico = analisar_grupo(df_problematico, "Liquida√ß√£o Problem√°tica (VP<VN)")

df_comparativo = pd.DataFrame([resultado_saudavel, resultado_problematico])

# 5. Apresentar a tabela comparativa e a conclus√£o final
print("\n--- TABELA COMPARATIVA FINAL ---")
display(df_comparativo.style.format({
    'N¬∫ Contratos √önicos': '{:,.0f}',
    'Valor Presente (R$)': 'R$ {:,.2f}',
    'PDD Total (R$)': 'R$ {:,.2f}',
    '% PDD': '{:,.2f}%'
}))

print("\n" + "-"*80)
print("CONCLUS√ÉO DA DEPURA√á√ÉO:")
if not df_comparativo.empty and df_comparativo.loc[1, '% PDD'] > df_comparativo.loc[0, '% PDD']:
    print("A TEORIA EST√Å 100% CONFIRMADA.")
    print("O sistema de fato registra dois eventos distintos no mesmo dia:")
    print("1. Liquida√ß√µes 'saud√°veis' com PDD baixo, onde o VP √© igual ao VN.")
    print("2. Liquida√ß√µes 'problem√°ticas' com PDD alt√≠ssimo, onde o VP √© menor que o VN, refletindo o risco.")
    print("N√£o h√° erro no script; esta √© a l√≥gica de neg√≥cio do sistema de origem.")
else:
    print("A teoria n√£o p√¥de ser confirmada. Os dois grupos n√£o apresentam a diferen√ßa esperada no PDD.")

print("#"*80 + "\n")
```

```python
# -*- coding: utf-8 -*-
"""
## An√°lise da Carteira de receb√≠veis (Vers√£o Final)

Script completo para an√°lise de estoque, com todas as corre√ß√µes de l√≥gica
implementadas, incluindo o ajuste no filtro de data para o c√°lculo da TIR.
"""

#* ==============>>>>>>>>
#* Bibliotecas >>>>>>>>
#* ==============>>>>>>>>
import pandas as pd
import numpy as np
import os
from scipy.optimize import brentq
import base64
from IPython.display import display

pd.options.display.max_columns = 100
pd.options.display.max_rows = 200

"""##### <span style="color:#CFFFE5;">Par√¢metro de Neg√≥cio</span>"""

DIAS_ATRASO_DEFINICAO_VENCIDO = 1

"""### <span style="color:#AEE5F9;"> Leitura e Prepara√ß√£o dos Dados
<span style="color: #FFB3B3; font-size: 15px; font-weight: bold;">
    ATEN√á√ÉO: REDIFINIR AQUI OS PATHS E NOMES DE ARQUIVOS, SE NECESS√ÅRIO.
</span>
"""

# =============================================================================
# LER DADOS =================================================================
# =============================================================================

#! PATHS ----------------------------------------------------------------------
DIRETORIO_DADOS = r'C:\Users\Leo\Desktop\Porto_Real\portoauto\src\vortx_estoques\data\131752-Estoque'
# ! ATEN√á√ÉO: Verifique se os nomes dos arquivos abaixo correspondem EXATAMENTE
# ! aos nomes na sua pasta, incluindo os espa√ßos.
ARQUIVOS_CSV = [
    'FIDC FCT CONSIGNADO II  - Estoque 05.09.25-Parte1.csv',
    'FIDC FCT CONSIGNADO II  - Estoque 05.09.25-Parte2.csv',
    'FIDC FCT CONSIGNADO II  - Estoque 05.09.25-Parte3.csv'
]
caminho_feriados = r'C:\Users\Leo\Desktop\Porto_Real\portoauto\feriados_nacionais.xls'
logo_path = r'C:\Users\Leo\Desktop\Porto_Real\portoauto\images\logo_inv.png'
output_path = r'C:\Users\Leo\Desktop\Porto_Real\portoauto\src\vortx_estoques\output'
#! ----------------------------------------------------------------------------

# --- Carga e unifica√ß√£o dos arquivos ---
lista_dfs = []
caminhos_completos = [os.path.join(DIRETORIO_DADOS, arq) for arq in ARQUIVOS_CSV]
for caminho in caminhos_completos:
    df_temp = pd.read_csv(caminho, sep=';', decimal=',', dayfirst=True, encoding='latin1')
    lista_dfs.append(df_temp)
df_merged = pd.concat(lista_dfs, ignore_index=True)
print(f"Dados unificados com sucesso. Total de {df_merged.shape[0]} registros carregados.")

# --- Limpeza e Prepara√ß√£o dos Dados ---
print("Iniciando limpeza e prepara√ß√£o dos dados...")
colunas_para_remover = [
    'NumeroBoleto', 'CampoAdicional1', 'CampoAdicional2', 'CampoAdicional3', 'CampoAdicional4',
    'CampoAdicional5', 'Registradora', 'IdContratoRegistradora', 'IdTituloRegistradora',
    'Coobricacao', 'PES_TIPO_PESSOA', 'TIT_CEDENTE_ENT_CODIGO', 'Cnae',
    'SecaoCNAEDescricao', 'NotaPdd', 'SAC_TIPO_PESSOA'
]
df_merged.drop(columns=colunas_para_remover, inplace=True, errors='ignore')

# Renomear colunas para compatibilidade
df_merged = df_merged.rename(columns={'IdTituloVortxOriginador': 'CCB', 'SacadoCnpjCpf': 'SacadoID'})

# Convers√£o de tipos de dados
cols_data = ['DataGeracao', 'DataAquisicao', 'DataVencimento', 'DataEmissao']
for col in cols_data:
    df_merged[col] = pd.to_datetime(df_merged[col], errors='coerce', dayfirst=True)

df_final2 = df_merged.copy()
del df_merged, lista_dfs
print(f"DataFrame pronto para an√°lise com {df_final2.shape[0]} linhas.")


"""### <span style="color:#AEE5F9;"> Colunas Auxiliares e Novas Segmenta√ß√µes

Cria√ß√£o de colunas derivadas para an√°lise de risco e novas segmenta√ß√µes de neg√≥cio.
"""

# =============================================================================
# Colunas Auxiliares =======================================================
# =============================================================================

data_referencia_relatorio = df_final2['DataGeracao'].max()
print(f"Usando a data do relat√≥rio como refer√™ncia para inadimpl√™ncia: {data_referencia_relatorio.strftime('%d/%m/%Y')}")

df_final2['_ValorLiquido'] = df_final2['ValorPresente'] - df_final2['PDDTotal']

dias_de_atraso = (data_referencia_relatorio - df_final2['DataVencimento']).dt.days
df_final2['_ParcelaVencida_Flag'] = ((dias_de_atraso >= DIAS_ATRASO_DEFINICAO_VENCIDO) & (dias_de_atraso > 0)).astype(int)
contratos_com_parcela_vencida = df_final2.groupby('CCB')['_ParcelaVencida_Flag'].max()
lista_ccbs_vencidos = contratos_com_parcela_vencida[contratos_com_parcela_vencida == 1].index
df_final2['_ContratoVencido_Flag'] = df_final2['CCB'].isin(lista_ccbs_vencidos).astype(int)

df_final2['_FlagDesagio'] = np.where(df_final2['ValorAquisicao'] > df_final2['ValorNominal'], '√Ågio', 'Des√°gio')

print("Cria√ß√£o de colunas auxiliares conclu√≠da.")
df_report = df_final2.copy()
del df_final2

"""### <span style="color:#AEE5F9;"> Qualidade e Consist√™ncia dos Dados

Verifico inconsist√™ncias, dados faltantes e caracter√≠sticas gerais da carteira.
"""
# =============================================================================
# Qualidade e Consist√™ncia dos Dados ==========================================
# =============================================================================
print("\n" + "="*80)
print("INICIANDO VERIFICA√á√ïES DE SANIDADE E QUALIDADE DOS DADOS")
print("="*80)
checks_results = {}
valor_presente_formatado = f"R$ {df_report['ValorPresente'].sum():,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.')
total_registros_formatado = f"{len(df_report):,}".replace(',', '.')
clientes_unicos_formatado = f"{df_report['SacadoID'].nunique():,}".replace(',', '.')
ccbs_unicos_formatado = f"{df_report['CCB'].nunique():,}".replace(',', '.')
checks_results['N√∫mero Total de Registros'] = total_registros_formatado
checks_results['Valor Presente Total da Carteira'] = valor_presente_formatado
checks_results['Per√≠odo da Carteira (Data de Aquisi√ß√£o)'] = f"{df_report['DataAquisicao'].min().strftime('%d/%m/%Y')} a {df_report['DataAquisicao'].max().strftime('%d/%m/%Y')}"
checks_results['N√∫mero de Clientes √önicos'] = clientes_unicos_formatado
checks_results['N√∫mero de Contratos √önicos'] = ccbs_unicos_formatado
checks_results['Valores Monet√°rios Negativos'] = (df_report[['ValorAquisicao', 'ValorNominal', 'ValorPresente', 'PDDTotal']] < 0).any().any()
checks_results['VP > Valor Nominal'] = (df_report['ValorPresente'] > df_report['ValorNominal']).sum()
checks_results['Valor Aquisi√ß√£o > Valor Nominal'] = (df_report['ValorAquisicao'] > df_report['ValorNominal']).sum()
checks_results['Data Aquisi√ß√£o > Data Vencimento'] = (df_report['DataAquisicao'] > df_report['DataVencimento']).sum()
print("Verifica√ß√µes de sanidade conclu√≠das.")


"""### <span style="color:#AEE5F9;"> Gera√ß√£o do Relat√≥rio Final

Consolida√ß√£o de todas as an√°lises em um relat√≥rio HTML interativo.
"""
# =============================================================================
# Gera√ß√£o do Relat√≥rio ======================================================
# =============================================================================
dimensoes_analise = {
    'Cedentes': 'CedenteNome',
    'Tipo de Ativo': 'TipoAtivo',
    'Situa√ß√£o': 'Situacao',
    'Pagamento Parcial': 'PagamentoParcial',
    'Flag √Ågio/Des√°gio': '_FlagDesagio'
}
dimensoes_analise = {k: v for k, v in dimensoes_analise.items() if v in df_report.columns}
os.makedirs(output_path, exist_ok=True)

#***********************
#* C√ÅLCULO M√âTRICAS
#***********************
print("\n" + "="*80)
print("INICIANDO C√ÅLCULO UNIFICADO DAS M√âTRICAS")
print("="*80)

dias_de_atraso = (data_referencia_relatorio - df_report['DataVencimento']).dt.days
df_report['_ValorVencido_1d'] = df_report.loc[dias_de_atraso >= 1, 'ValorPresente'].fillna(0)
df_report['_ValorVencido_30d'] = df_report.loc[dias_de_atraso >= 30, 'ValorPresente'].fillna(0)
df_report['_ValorVencido_60d'] = df_report.loc[dias_de_atraso >= 60, 'ValorPresente'].fillna(0)
df_report.fillna({'_ValorVencido_1d': 0, '_ValorVencido_30d': 0, '_ValorVencido_60d': 0}, inplace=True)

vp_col_name = 'Valor Presente \n(R$ MM)'
vl_col_name = 'Valor L√≠quido \n(R$ MM)'
col_contratos_venc_perc = f"% Contratos Venc. (>{DIAS_ATRASO_DEFINICAO_VENCIDO}d)"
tabelas_metricas = {}

for nome_analise, coluna in dimensoes_analise.items():
    print(f"Calculando m√©tricas para a dimens√£o: '{nome_analise}'...")
    grouped = df_report.groupby(coluna, observed=True)
    total_contratos_unicos = grouped['CCB'].nunique()
    contratos_vencidos_unicos = df_report[df_report['_ContratoVencido_Flag'] == 1].groupby(coluna, observed=True)['CCB'].nunique()
    df_metricas = pd.DataFrame({'N¬∫ Contratos √önicos': total_contratos_unicos})
    somas_financeiras = grouped[['_ValorLiquido', 'ValorPresente', '_ValorVencido_1d', '_ValorVencido_30d', '_ValorVencido_60d']].sum()
    df_metricas = df_metricas.join(somas_financeiras)
    df_metricas['%PDD'] = (1 - df_metricas['_ValorLiquido'] / df_metricas['ValorPresente']) * 100
    df_metricas[col_contratos_venc_perc] = (contratos_vencidos_unicos.reindex(df_metricas.index).fillna(0) / df_metricas['N¬∫ Contratos √önicos']).fillna(0) * 100
    df_metricas['Venc. 1d / VP'] = (df_metricas['_ValorVencido_1d'] / df_metricas['ValorPresente']) * 100
    df_metricas['Venc. 30d / VP'] = (df_metricas['_ValorVencido_30d'] / df_metricas['ValorPresente']) * 100
    df_metricas['Venc. 60d / VP'] = (df_metricas['_ValorVencido_60d'] / df_metricas['ValorPresente']) * 100
    df_metricas = df_metricas.rename(columns={'ValorPresente': vp_col_name, '_ValorLiquido': vl_col_name, 'N¬∫ Contratos √önicos': 'N¬∫ Contratos'})
    df_metricas[[vp_col_name, vl_col_name]] /= 1e6
    df_metricas = df_metricas.drop(columns=['_ValorLiquido', '_ValorVencido_1d', '_ValorVencido_30d', '_ValorVencido_60d'], errors='ignore')
    tabelas_metricas[nome_analise] = df_metricas
print("C√°lculo unificado de m√©tricas conclu√≠do.")

#***********************
#* TICKET M√âDIO
#***********************
tabelas_ticket = {}
for nome_analise, coluna in dimensoes_analise.items():
    df_temp = df_report.dropna(subset=[coluna, 'ValorPresente', 'ValorNominal'])
    if df_temp.empty: continue
    grouped = df_temp.groupby(coluna, observed=True)
    numerador = grouped.apply(lambda g: (g['ValorNominal'] * g['ValorPresente']).sum(), include_groups=False)
    denominador = grouped['ValorPresente'].sum()
    ticket_ponderado = (numerador / denominador).replace([np.inf, -np.inf], 0)
    ticket_ponderado.name = "Ticket Ponderado (R$)"
    tabelas_ticket[nome_analise] = pd.DataFrame(ticket_ponderado)
print("Ticket m√©dio calculado.")


# << NOVO BLOCO: C√ÅLCULO DAS M√âTRICAS ADICIONAIS ANTES DO LOOP PRINCIPAL >>
#***********************
#* REPRESENTATIVIDADE DA TIR
#***********************
print("Calculando representatividade da amostra para a TIR...")
tabelas_representatividade = {}
df_avencer_temp = df_report[df_report['DataVencimento'] >= data_referencia_relatorio]

for nome_analise, coluna in dimensoes_analise.items():
    if coluna not in df_report.columns: continue
    
    total_contratos = df_report.groupby(coluna, observed=True)['CCB'].nunique()
    contratos_com_futuro = df_avencer_temp.groupby(coluna, observed=True)['CCB'].nunique()
    
    df_rep = pd.DataFrame({'Total Contratos': total_contratos, 'Amostra TIR (Contratos)': contratos_com_futuro}).fillna(0)
    df_rep['Amostra TIR (Contratos)'] = df_rep['Amostra TIR (Contratos)'].astype(int)
    
    percentual = (df_rep['Amostra TIR (Contratos)'] / df_rep['Total Contratos'] * 100).fillna(0)
    df_rep['Amostra TIR (Contratos)'] = df_rep['Amostra TIR (Contratos)'].astype(str) + ' (' + percentual.map('{:,.2f}%'.format) + ')'
    
    tabelas_representatividade[nome_analise] = df_rep[['Amostra TIR (Contratos)']]
print("C√°lculo de representatividade conclu√≠do.")

#***********************
#* DATA MEDIANA DE VENCIMENTO
#***********************
print("Calculando a data mediana de vencimento...")
tabelas_data_mediana = {}

for nome_analise, coluna in dimensoes_analise.items():
    if coluna not in df_report.columns: continue
    
    medianas = df_avencer_temp.groupby(coluna, observed=True)['DataVencimento'].median()
    df_medianas = pd.DataFrame(medianas)
    df_medianas['DataVencimento'] = df_medianas['DataVencimento'].dt.strftime('%d/%m/%Y').fillna('-')
    df_medianas = df_medianas.rename(columns={'DataVencimento': 'Data Mediana Venc.'})
    tabelas_data_mediana[nome_analise] = df_medianas
print("C√°lculo da data mediana conclu√≠do.")


"""#### <span style="color:#FFDAC1;"> C√°lculo da TIR

C√°lculo da TIR (XIRR) para a carteira a vencer. Como n√£o h√° dados de custo, calculamos a TIR Bruta e a TIR L√≠quida de PDD.
"""
#******************
#* TIR com brentq
#*******************
def calculate_xirr(cash_flows, days):
    cash_flows = np.array(cash_flows)
    days = np.array(days)
    def npv(rate):
        if rate <= -1: return float('inf')
        with np.errstate(divide='ignore', over='ignore'):
            return np.sum(cash_flows / (1 + rate) ** (days / 21.0))
    try:
        return brentq(npv, 0, 1.0)
    except (RuntimeError, ValueError):
        return np.nan

print("\n" + "="*80)
print("INICIANDO C√ÅLCULO DA TAXA INTERNA DE RETORNO (TIR)")
print("="*80)

ref_date = df_report['DataGeracao'].max()
print(f"Data de Refer√™ncia para o c√°lculo da TIR: {ref_date.strftime('%d/%m/%Y')}")

try:
    df_feriados = pd.read_excel(caminho_feriados)
    holidays = pd.to_datetime(df_feriados['Data'], errors='coerce').dropna().dt.date.values
    print(f"Sucesso: {len(holidays)} feriados carregados.")
except Exception as e:
    print(f"[AVISO] N√£o foi poss√≠vel carregar feriados: {e}. O c√°lculo de dias √∫teis pode ser impreciso.")
    holidays = []

df_avencer = df_report[df_report['DataVencimento'] >= ref_date].copy()


if not df_avencer.empty:
    try:
        start_date_np = np.datetime64(ref_date.date(), 'D')
        end_dates_np = df_avencer['DataVencimento'].dt.date.values.astype('datetime64[D]')
        holidays_np = np.array(holidays, dtype='datetime64[D]')

        df_avencer.loc[:, '_DIAS_UTEIS_'] = np.busday_count(start_date_np, end_dates_np, holidays=holidays_np)
        
        print("C√°lculo de dias √∫teis realizado com sucesso.")
    except Exception as e:
        print(f"[ERRO] Falha ao calcular dias √∫teis: {e}")
        df_avencer.loc[:, '_DIAS_UTEIS_'] = np.nan
else:
    print("[AVISO] N√£o h√° parcelas futuras ('a vencer') na base de dados. O c√°lculo de TIR ser√° pulado.")
    df_avencer['_DIAS_UTEIS_'] = np.nan


all_tirs = []
cat_cols = list(dimensoes_analise.values())
segmentos_para_analise = [('Carteira Total', 'Todos')] + \
                          [(col, seg) for col in cat_cols if col in df_avencer.columns for seg in df_avencer[col].dropna().unique()]

for tipo_dimensao, segmento in segmentos_para_analise:
    if tipo_dimensao == 'Carteira Total':
        df_segmento = df_avencer.copy()
    else:
        df_segmento = df_avencer[df_avencer[tipo_dimensao] == segmento].copy()

    if df_segmento.empty or df_segmento['_DIAS_UTEIS_'].isnull().all():
        continue

    vp_bruto = df_segmento['ValorPresente'].sum()
    tir_bruta, tir_pdd = np.nan, np.nan

    if vp_bruto > 0:
        fluxos_brutos = df_segmento.groupby('_DIAS_UTEIS_')['ValorNominal'].sum()
        tir_bruta = calculate_xirr([-vp_bruto] + fluxos_brutos.values.tolist(), [0] + fluxos_brutos.index.tolist())

        pdd_total_segmento = df_segmento['PDDTotal'].sum()
        pdd_rate = (pdd_total_segmento / vp_bruto) if pd.notna(pdd_total_segmento) and vp_bruto > 0 else 0.0
        df_segmento['Fluxo_PDD'] = df_segmento['ValorNominal'] * (1 - pdd_rate)
        fluxos_pdd = df_segmento.groupby('_DIAS_UTEIS_')['Fluxo_PDD'].sum()
        tir_pdd = calculate_xirr([-vp_bruto] + fluxos_pdd.values.tolist(), [0] + fluxos_pdd.index.tolist())

    all_tirs.append({
        'DimensaoColuna': tipo_dimensao, 'Segmento': segmento,
        'Valor Presente TIR (M)': vp_bruto / 1e6,
        'TIR Bruta \n(% a.m. )': tir_bruta * 100 if pd.notna(tir_bruta) else np.nan,
        'TIR L√≠quida de PDD \n(% a.m. )': tir_pdd * 100 if pd.notna(tir_pdd) else np.nan
    })

df_tir_summary = pd.DataFrame(all_tirs)
if not df_tir_summary.empty:
    tir_cols_to_fill = [col for col in df_tir_summary.columns if 'TIR' in col]
    df_tir_summary[tir_cols_to_fill] = df_tir_summary[tir_cols_to_fill].fillna(-100.0)

print("C√°lculo de TIR conclu√≠do.")

"""#### <span style="color:#FFDAC1;"> Montagem do HTML

Uno todos os elementos em um relat√≥rio HTML interativo.
"""
#***********************
#* GERA√á√ÉO DO RELAT√ìRIO HTML
#***********************

print("\n" + "="*80)
print("GERANDO RELAT√ìRIO HTML FINAL")
print("="*80)

def encode_image_to_base64(image_path):
    try:
        with open(image_path, "rb") as image_file:
            return base64.b64encode(image_file.read()).decode('utf-8')
    except FileNotFoundError:
        print(f"[ATEN√á√ÉO] Arquivo de imagem n√£o encontrado em: {image_path}. A logo n√£o ser√° exibida.")
        return None

logo_base64 = encode_image_to_base64(logo_path)
report_date = ref_date.strftime('%d/%m/%Y')

html_css = """
<style>
    .single-entry-message { background-color: #eef7f7; border-left: 5px solid #76c6c5; padding: 15px 20px; margin: 10px 0; font-size: 1.0em; color: #313131; }
    .single-entry-message strong { color: #163f3f; }
    .summary-title { font-size: 1.4em; color: #163f3f; border-bottom: 2px solid #76c6c5; padding-bottom: 10px; margin-top: 20px; margin-bottom: 20px; }
    .checks-container { display: grid; grid-template-columns: repeat(auto-fit, minmax(350px, 1fr)); gap: 15px; margin-bottom: 20px; }
    .check-item { background-color: #f5f5f5; padding: 12px; border-radius: 5px; border-left: 4px solid #76c6c5; font-size: 0.95em; }
    .check-item strong { color: #163f3f; }
    .dataframe-checks th { background-color: #5b8c8c; }
    .check-table-wrapper h4 { margin-top: 20px; margin-bottom: 10px; color: #163f3f; border-bottom: 2px solid #eeeeee; padding-bottom: 5px; }
    body { font-family: "Gill Sans MT", Arial, sans-serif; background-color: #f9f9f9; color: #313131; margin: 0; padding: 0; }
    .main-content { padding: 25px; }
    header { background-color: #163f3f; color: #FFFFFF; padding: 20px 40px; display: flex; justify-content: space-between; align-items: center; border-bottom: 5px solid #76c6c5; }
    header .logo img { height: 75px; }
    header .report-title { text-align: left; font-family: "Gill Sans MT", Arial, sans-serif; }
    header .report-title h1, header .report-title h2, header .report-title h3 { margin: 0; padding: 0; font-weight: normal; }
    header .report-title h1 { font-size: 1.6em; }
    header .report-title h2 { font-size: 1.4em; color: #d0d0d0; }
    header .report-title h3 { font-size: 1.1em; color: #a0a0a0; }
    .container-botoes { display: flex; flex-wrap: wrap; gap: 15px; margin-bottom: 25px; margin-top: 40px; }
    .container-botoes > details { flex: 1 1 280px; border: 1px solid #76c6c5; border-radius: 8px; overflow: hidden; }
    .container-botoes > details[open] { flex-basis: 100%; }
    details summary { font-size: 1.1em; font-weight: bold; color: #FFFFFF; background-color: #163f3f; padding: 15px 20px; cursor: pointer; outline: none; list-style-type: none; }
    details summary:hover { background-color: #0e5d5f; }
    details[open] summary { background-color: #76c6c5; color: #313131; }
    details[open] summary:hover { filter: brightness(95%); }
    summary::-webkit-details-marker { display: none; }
    summary::before { content: '‚ñ∫ '; margin-right: 8px; font-size: 0.8em;}
    details[open] summary::before { content: '‚ñº '; }
    .content-wrapper { padding: 20px; background-color: #FFFFFF; overflow-x: auto; border: 1px solid #ddd; border-radius: 8px; }
    table.dataframe, th, td { border: 1px solid #bbbbbb; }
    table.dataframe { border-collapse: collapse; width: 100%; }
    th, td { text-align: left; padding: 10px; vertical-align: middle; }
    th { background-color: #163f3f; color: #FFFFFF; }
    tr:nth-child(even) { background-color: #eeeeee; }
    footer { background-color: #f0f0f0; color: #555555; font-size: 0.8em; line-height: 1.6; padding: 25px 40px; margin-top: 40px; border-top: 1px solid #dddddd; }
    footer .disclaimer { margin-top: 20px; font-style: italic; border-top: 1px solid #dddddd; padding-top: 20px; }
</style>
"""

html_parts = ["<!DOCTYPE html><html lang='pt-BR'><head>",
              "<meta charset='UTF-8'><title>An√°lise de Desempenho - FCT Consignado II</title>",
              html_css, "</head><body>"]

html_parts.append(f"<header><div class='report-title'><h1>An√°lise de desempenho</h1><h2>FCT CONSIGNADO II</h2><h3>{report_date}</h3></div>")
if logo_base64:
    html_parts.append(f'<div class="logo"><img src="data:image/png;base64,{logo_base64}" alt="Logo"></div>')
html_parts.append("</header><div class='main-content'>")

mapa_descricoes = {
    'Cedentes': 'Analisa as m√©tricas de risco e retorno agrupadas por cada Cedente.',
    'Tipo de Ativo': 'Compara o desempenho entre os diferentes tipos de ativos (CCB, CT - Contrato).',
    'Situa√ß√£o': 'Compara o desempenho dos t√≠tulos com base na sua situa√ß√£o atual (Ex: Aditado).',
    'Pagamento Parcial': 'Verifica se h√° impacto nas m√©tricas para t√≠tulos que aceitam pagamento parcial.',
    'Flag √Ågio/Des√°gio': 'Compara o desempenho de contratos adquiridos com √°gio vs. des√°gio.'
}

# Verifica√ß√µes de Sanidade dos Dados
html_parts.append("<details open><summary>Verifica√ß√µes e Sanidade dos Dados</summary><div class='content-wrapper'>")
simple_checks_html = "<div class='checks-container'>"
table_checks_html = "<div class='check-table-wrapper'>"
for key, value in checks_results.items():
    if '[TABELA]' in key:
        clean_key = key.replace('[TABELA]', '').strip()
        table_checks_html += f"<h4>{clean_key}</h4>{str(value)}"
    else:
        simple_checks_html += f"<div class='check-item'><strong>{key}:</strong> {value}</div>"
simple_checks_html += "</div>"
table_checks_html += "</div>"
html_parts.append(simple_checks_html)
html_parts.append(table_checks_html)
html_parts.append("</div></details>")

# << ALTERADO: Bloco do Resumo da Carteira Total agora tamb√©m inclui as novas m√©tricas >>
# Resumo da Carteira Total
print("--> Gerando tabela de resumo da carteira total...")
df_avencer_total = df_report[df_report['DataVencimento'] >= data_referencia_relatorio]
data_mediana_total = df_avencer_total['DataVencimento'].median()

total_metrics = {
    'N¬∫ Contratos': df_report['CCB'].nunique(),
    'Amostra TIR (Contratos)': f"{df_avencer_total['CCB'].nunique():,} ({df_avencer_total['CCB'].nunique() / df_report['CCB'].nunique() * 100:.2f}%)",
    'Data Mediana Venc.': data_mediana_total.strftime('%d/%m/%Y') if pd.notna(data_mediana_total) else '-',
    col_contratos_venc_perc: (df_report[df_report['_ContratoVencido_Flag'] == 1]['CCB'].nunique() / df_report['CCB'].nunique()) * 100,
    vl_col_name: df_report['_ValorLiquido'].sum() / 1e6,
    vp_col_name: df_report['ValorPresente'].sum() / 1e6,
    '%PDD': (1 - df_report['_ValorLiquido'].sum() / df_report['ValorPresente'].sum()) * 100 if df_report['ValorPresente'].sum() > 0 else 0,
    'Venc. 1d / VP': (df_report['_ValorVencido_1d'].sum() / df_report['ValorPresente'].sum()) * 100 if df_report['ValorPresente'].sum() > 0 else 0,
    'Venc. 30d / VP': (df_report['_ValorVencido_30d'].sum() / df_report['ValorPresente'].sum()) * 100 if df_report['ValorPresente'].sum() > 0 else 0,
    'Venc. 60d / VP': (df_report['_ValorVencido_60d'].sum() / df_report['ValorPresente'].sum()) * 100 if df_report['ValorPresente'].sum() > 0 else 0,
    'Ticket Ponderado (R$)': (df_report['ValorNominal'] * df_report['ValorPresente']).sum() / df_report['ValorPresente'].sum() if df_report['ValorPresente'].sum() > 0 else 0,
}
df_total = pd.DataFrame([total_metrics], index=['Total'])
if not df_tir_summary.empty:
    df_tir_total = df_tir_summary[df_tir_summary['DimensaoColuna'] == 'Carteira Total'].drop(columns=['DimensaoColuna', 'Segmento', 'Valor Presente TIR (M)'])
    if not df_tir_total.empty:
        df_tir_total.index = ['Total']
        df_total = pd.concat([df_total, df_tir_total], axis=1)

# << ALTERADO: Ordem das colunas do resumo total >>
colunas_vencimento = ['Venc. 1d / VP', 'Venc. 30d / VP', 'Venc. 60d / VP']
colunas_ordem_total = ['N¬∫ Contratos', 'Amostra TIR (Contratos)', 'Data Mediana Venc.', col_contratos_venc_perc, vl_col_name, vp_col_name, '%PDD'] + colunas_vencimento + ['Ticket Ponderado (R$)']
ordem_ideal_tir = ['TIR Bruta \n(% a.m. )', 'TIR L√≠quida de PDD \n(% a.m. )']
colunas_tir_ordenadas = [col for col in ordem_ideal_tir if col in df_total.columns]
colunas_finais_total = [col for col in colunas_ordem_total + colunas_tir_ordenadas if col in df_total.columns]
df_total = df_total[colunas_finais_total]

formatters = {
    vl_col_name: '{:,.2f}'.format, vp_col_name: '{:,.2f}'.format,
    'N¬∫ Contratos': lambda x: f'{x:,.0f}'.replace(',', '.'),
    'Ticket Ponderado (R$)': lambda x: f'R$ {x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'),
    '%PDD': '{:,.2f}%'.format
}
formatters[col_contratos_venc_perc] = '{:,.2f}%'.format
for col in colunas_vencimento: formatters[col] = '{:,.2f}%'.format
for col in colunas_tir_ordenadas: formatters[col] = '{:,.2f}%'.format

df_total.columns = [col.replace('\n', '<br>') for col in df_total.columns]

html_parts.append("<h2 class='summary-title'>Resumo da Carteira Total</h2><div class='content-wrapper'>")
html_parts.append(df_total.to_html(index=False, classes='dataframe', formatters=formatters, na_rep='-', escape=False))
html_parts.append("</div>")


# << ALTERADO: Loop principal agora junta as novas tabelas e exibe as novas colunas >>
# In√≠cio dos bot√µes retr√°teis
html_parts.append("<div class='container-botoes'>")
for nome_analise, coluna in dimensoes_analise.items():
    if coluna not in df_report.columns or df_report[coluna].isnull().all(): continue
    print(f"--> Processando e gerando HTML para: '{nome_analise}'")

    df_final = tabelas_metricas.get(nome_analise)
    df_ticket = tabelas_ticket.get(nome_analise, pd.DataFrame())
    df_rep = tabelas_representatividade.get(nome_analise, pd.DataFrame())
    df_mediana = tabelas_data_mediana.get(nome_analise, pd.DataFrame())

    # Juntando todas as tabelas
    df_final = df_final.join(df_rep, how='outer')
    df_final = df_final.join(df_mediana, how='outer')
    df_final = df_final.join(df_ticket, how='outer')
    if not df_tir_summary.empty:
        df_tir = df_tir_summary[df_tir_summary['DimensaoColuna'] == coluna].set_index('Segmento')
        if not df_tir.empty:
            df_final = df_final.join(df_tir.drop(columns=['DimensaoColuna', 'Valor Presente TIR (M)']), how='outer')

    df_final.index.name = nome_analise
    df_final.reset_index(inplace=True)

    # Nova ordem de colunas
    colunas_ordem = [
        nome_analise, 'N¬∫ Contratos', 'Amostra TIR (Contratos)', 'Data Mediana Venc.', 
        col_contratos_venc_perc, vl_col_name, vp_col_name, '%PDD'
    ] + colunas_vencimento
    
    if 'Ticket Ponderado (R$)' in df_final.columns: colunas_ordem.append('Ticket Ponderado (R$)')

    colunas_finais = [col for col in colunas_ordem + colunas_tir_ordenadas if col in df_final.columns]
    outras_colunas = [col for col in df_final.columns if col not in df_final.columns]
    df_final = df_final[colunas_finais + outras_colunas]

    df_final = df_final.sort_values(vp_col_name, ascending=False).reset_index(drop=True)
    df_final.columns = [col.replace('\n', '<br>') for col in df_final.columns]

    html_parts.append("<details>")
    descricao = mapa_descricoes.get(nome_analise, 'Descri√ß√£o n√£o dispon√≠vel.')
    html_parts.append(f'<summary title="{descricao}">{nome_analise}</summary>')
    html_parts.append("<div class='content-wrapper'>")
    html_table = df_final.to_html(index=False, classes='dataframe', formatters=formatters, na_rep='-', escape=False)
    html_parts.append(html_table)
    html_parts.append("</div></details>")

html_parts.append("</div></div>")

footer_main_text = "Este documento tem como objetivo apresentar uma an√°lise de desempenho do fundo FCT Consignado II (CNPJ 52.203.615/0001-19)..."
footer_disclaimer = "Disclaimer: Este relat√≥rio foi preparado pelo Porto Real Investimentos exclusivamente para fins informativos..."
html_parts.append(f"<footer><p>{footer_main_text.strip()}</p><div class='disclaimer'>{footer_disclaimer.strip()}</div></footer>")
html_parts.append("</body></html>")

final_html_content = "\n".join(html_parts)
html_output_filename = os.path.join(output_path, 'analise_carteira_estoque.html')
try:
    with open(html_output_filename, 'w', encoding='utf-8') as f:
        f.write(final_html_content)
    print("\n" + "="*80)
    print("AN√ÅLISE CONCLU√çDA COM SUCESSO!")
    print(f"O relat√≥rio HTML final foi salvo em: {html_output_filename}")
    print("="*80)
except Exception as e:
    print(f"\n[ERRO GRAVE] N√£o foi poss√≠vel salvar o arquivo HTML: {e}")

#del df_report, df_total
```

```python
# -*- coding: utf-8 -*-
"""
## An√°lise da Carteira de receb√≠veis (Vers√£o Final)

Script completo para an√°lise de estoque, com todas as corre√ß√µes de l√≥gica
implementadas, incluindo o ajuste no filtro de data para o c√°lculo da TIR.
"""

#* ==============>>>>>>>>
#* Bibliotecas >>>>>>>>
#* ==============>>>>>>>>
import pandas as pd #* para manipula√ß√£o e an√°lise de dados
import numpy as np #* para opera√ß√µes num√©ricas, especialmente em arrays
import os #* para interagir com o sistema operacional (manipular caminhos de arquivos)
from scipy.optimize import brentq #* para encontrar a raiz de fun√ß√µes (usado no c√°lculo da TIR)
import base64 #* para codificar a imagem do logo para o relat√≥rio HTML
from IPython.display import display #* para melhor visualiza√ß√£o de dataframes no Jupyter

#? Configura√ß√µes do Pandas para exibir mais colunas e linhas
pd.options.display.max_columns = 100
pd.options.display.max_rows = 200

"""##### <span style="color:#CFFFE5;">Par√¢metro de Neg√≥cio</span>"""
"""DIAS_ATRASO_DEFINICAO_VENCIDO: int
Define o n√∫mero m√≠nimo de dias de atraso para que uma parcela seja
considerada 'vencida' na an√°lise.
"""
DIAS_ATRASO_DEFINICAO_VENCIDO = 1

"""### <span style="color:#AEE5F9;"> Leitura e Prepara√ß√£o dos Dados
<span style="color: #FFB3B3; font-size: 15px; font-weight: bold;">
    ATEN√á√ÉO: REDIFINIR AQUI OS PATHS E NOMES DE ARQUIVOS, SE NECESS√ÅRIO.
</span>
"""

# =============================================================================
# LER DADOS =================================================================
# =============================================================================

#! PATHS ----------------------------------------------------------------------
"""DIRETORIO_DADOS: str
Caminho para a pasta que cont√©m os arquivos CSV de estoque.
"""
DIRETORIO_DADOS = r'C:\Users\Leo\Desktop\Porto_Real\portoauto\src\vortx_estoques\data\131752-Estoque'

"""ARQUIVOS_CSV: list
Lista com os nomes dos arquivos CSV a serem lidos e concatenados.
#! ATEN√á√ÉO: Verifique se os nomes dos arquivos abaixo correspondem EXATAMENTE
#! aos nomes na sua pasta, incluindo os espa√ßos.
"""
ARQUIVOS_CSV = [
    'FIDC FCT CONSIGNADO II  - Estoque 05.09.25-Parte1.csv',
    'FIDC FCT CONSIGNADO II  - Estoque 05.09.25-Parte2.csv',
    'FIDC FCT CONSIGNADO II  - Estoque 05.09.25-Parte3.csv'
]
"""caminho_feriados: str
Caminho para o arquivo Excel contendo a lista de feriados nacionais.
"""
caminho_feriados = r'C:\Users\Leo\Desktop\Porto_Real\portoauto\feriados_nacionais.xls'

"""logo_path: str
Caminho para a imagem do logo que ser√° exibida no relat√≥rio HTML.
"""
logo_path = r'C:\Users\Leo\Desktop\Porto_Real\portoauto\images\logo_inv.png'

"""output_path: str
Caminho para a pasta onde o relat√≥rio HTML final ser√° salvo.
"""
output_path = r'C:\Users\Leo\Desktop\Porto_Real\portoauto\src\vortx_estoques\output'
#! ----------------------------------------------------------------------------

# --- Carga e unifica√ß√£o dos arquivos ---
lista_dfs = [] #? lista tempor√°ria para armazenar os dataframes de cada parte do relat√≥rio
caminhos_completos = [os.path.join(DIRETORIO_DADOS, arq) for arq in ARQUIVOS_CSV]
#* cria o caminho completo para cada arquivo

for caminho in caminhos_completos:
    #* l√™ cada arquivo csv, especificando o separador, decimal, formato de data e encoding
    df_temp = pd.read_csv(caminho, sep=';', decimal=',', dayfirst=True, encoding='latin1')
    lista_dfs.append(df_temp)
df_merged = pd.concat(lista_dfs, ignore_index=True) #? une todos os dataframes da lista em um s√≥, resetando o √≠ndice

print(f"Dados unificados com sucesso. Total de {df_merged.shape[0]} registros carregados.")

# --- Limpeza e Prepara√ß√£o dos Dados ---
print("Iniciando limpeza e prepara√ß√£o dos dados...")
"""colunas_para_remover: list
Lista de colunas que n√£o s√£o relevantes para a an√°lise e ser√£o removidas
para simplificar o dataframe.
"""
colunas_para_remover = [
    'NumeroBoleto', 'CampoAdicional1', 'CampoAdicional2', 'CampoAdicional3', 'CampoAdicional4',
    'CampoAdicional5', 'Registradora', 'IdContratoRegistradora', 'IdTituloRegistradora',
    'Coobricacao', 'PES_TIPO_PESSOA', 'TIT_CEDENTE_ENT_CODIGO', 'Cnae',
    'SecaoCNAEDescricao', 'NotaPdd', 'SAC_TIPO_PESSOA'
]
df_merged.drop(columns=colunas_para_remover, inplace=True, errors='ignore') #? remove as colunas, ignorando erros se alguma n√£o existir

# Renomeia colunas para manter a compatibilidade e clareza no script
df_merged = df_merged.rename(columns={'IdTituloVortxOriginador': 'CCB', 'SacadoCnpjCpf': 'SacadoID'})

# Converte colunas de data para o formato datetime do pandas
cols_data = ['DataGeracao', 'DataAquisicao', 'DataVencimento', 'DataEmissao']
for col in cols_data:
    df_merged[col] = pd.to_datetime(df_merged[col], errors='coerce', dayfirst=True) 
    
    #? 'coerce' transforma datas invalidas em NaT (Not a Time)

df_final2 = df_merged.copy() #* cria uma c√≥pia para trabalhar, preservando o original
del df_merged, lista_dfs #? libera mem√≥ria deletando os dataframes que n√£o ser√£o mais usados
print(f"DataFrame pronto para an√°lise com {df_final2.shape[0]} linhas.")


"""### <span style="color:#AEE5F9;"> Colunas Auxiliares e Novas Segmenta√ß√µes

Cria√ß√£o de colunas derivadas para an√°lise de risco e novas segmenta√ß√µes de neg√≥cio.
"""

# =============================================================================
# Colunas Auxiliares =======================================================
# =============================================================================
"""data_referencia_relatorio: datetime
A data mais recente encontrada na coluna 'DataGeracao', usada como o "dia de hoje"
para todos os c√°lculos de vencimento e atraso.
"""
data_referencia_relatorio = df_final2['DataGeracao'].max()
print(f"Usando a data do relat√≥rio como refer√™ncia para inadimpl√™ncia: {data_referencia_relatorio.strftime('%d/%m/%Y')}")

"""_ValorLiquido: float
Representa o Valor Presente da parcela ap√≥s a dedu√ß√£o da Provis√£o para
Devedores Duvidosos (PDD). √â o valor 'l√≠quido' de risco de cr√©dito.
"""
df_final2['_ValorLiquido'] = df_final2['ValorPresente'] - df_final2['PDDTotal']

#* L√≥gica para identificar contratos vencidos:
# 1. Calcula os dias de atraso para CADA PARCELA.
dias_de_atraso = (data_referencia_relatorio - df_final2['DataVencimento']).dt.days


# 2. Marca cada PARCELA como vencida (1) ou n√£o (0) com base na regra de neg√≥cio.
df_final2['_ParcelaVencida_Flag'] = ((dias_de_atraso >= DIAS_ATRASO_DEFINICAO_VENCIDO) & (dias_de_atraso > 0)).astype(int)


# 3. Agrupa por contrato (CCB) e verifica se QUALQUER parcela est√° vencida (usando .max()).
contratos_com_parcela_vencida = df_final2.groupby('CCB')['_ParcelaVencida_Flag'].max()


# 4. Cria uma lista com os IDs dos contratos que t√™m pelo menos uma parcela vencida.
lista_ccbs_vencidos = contratos_com_parcela_vencida[contratos_com_parcela_vencida == 1].index


# 5. Cria a flag final a n√≠vel de CONTRATO, marcando todas as parcelas de um contrato vencido.
df_final2['_ContratoVencido_Flag'] = df_final2['CCB'].isin(lista_ccbs_vencidos).astype(int)



"""_FlagDesagio: str
Cria uma categoria para cada parcela, indicando se ela foi adquirida com
'√Ågio' (Valor de Aquisi√ß√£o > Valor Nominal) ou 'Des√°gio'.
"""
df_final2['_FlagDesagio'] = np.where(df_final2['ValorAquisicao'] > df_final2['ValorNominal'], '√Ågio', 'Des√°gio')

print("Cria√ß√£o de colunas auxiliares conclu√≠da.")
df_report = df_final2.copy() #? DataFrame final que ser√° usado para o relat√≥rio
del df_final2


"""### <span style="color:#AEE5F9;"> Qualidade e Consist√™ncia dos Dados

Verifico inconsist√™ncias, dados faltantes e caracter√≠sticas gerais da carteira.
"""
# =============================================================================
# Qualidade e Consist√™ncia dos Dados ==========================================
# =============================================================================
print("\n" + "="*80)
print("INICIANDO VERIFICA√á√ïES DE SANIDADE E QUALIDADE DOS DADOS")
print("="*80)
"""checks_results: dict
Um dicion√°rio para armazenar os resultados das verifica√ß√µes de qualidade.
Estes resultados ser√£o exibidos no in√≠cio do relat√≥rio HTML.
"""
checks_results = {}
#? formata os n√∫meros para o padr√£o brasileiro (milhar com ponto, decimal com v√≠rgula)
valor_presente_formatado = f"R$ {df_report['ValorPresente'].sum():,.2f}".replace(',', 'X').replace('.', ',').replace('X', '.')
total_registros_formatado = f"{len(df_report):,}".replace(',', '.')
clientes_unicos_formatado = f"{df_report['SacadoID'].nunique():,}".replace(',', '.')
ccbs_unicos_formatado = f"{df_report['CCB'].nunique():,}".replace(',', '.')

#* armazena as principais m√©tricas descritivas da carteira
checks_results['N√∫mero Total de Registros'] = total_registros_formatado
checks_results['Valor Presente Total da Carteira'] = valor_presente_formatado
checks_results['Per√≠odo da Carteira (Data de Aquisi√ß√£o)'] = f"{df_report['DataAquisicao'].min().strftime('%d/%m/%Y')} a {df_report['DataAquisicao'].max().strftime('%d/%m/%Y')}"
checks_results['N√∫mero de Clientes √önicos'] = clientes_unicos_formatado
checks_results['N√∫mero de Contratos √önicos'] = ccbs_unicos_formatado

#* verifica a integridade dos dados
checks_results['Valores Monet√°rios Negativos'] = (df_report[['ValorAquisicao', 'ValorNominal', 'ValorPresente', 'PDDTotal']] < 0).any().any()
checks_results['VP > Valor Nominal'] = (df_report['ValorPresente'] > df_report['ValorNominal']).sum()
checks_results['Valor Aquisi√ß√£o > Valor Nominal'] = (df_report['ValorAquisicao'] > df_report['ValorNominal']).sum()
checks_results['Data Aquisi√ß√£o > Data Vencimento'] = (df_report['DataAquisicao'] > df_report['DataVencimento']).sum()
print("Verifica√ß√µes de sanidade conclu√≠das.")


"""### <span style="color:#AEE5F9;"> Gera√ß√£o do Relat√≥rio Final

Consolida√ß√£o de todas as an√°lises em um relat√≥rio HTML interativo.
"""
# =============================================================================
# Gera√ß√£o do Relat√≥rio ======================================================
# =============================================================================
"""dimensoes_analise: dict
Dicion√°rio que mapeia os nomes das an√°lises (chaves) para as colunas
correspondentes no DataFrame (valores). O relat√≥rio ser√° segmentado por essas dimens√µes.
"""
dimensoes_analise = {
    'Cedentes': 'CedenteNome',
    'Tipo de Ativo': 'TipoAtivo',
    'Situa√ß√£o': 'Situacao',
    'Pagamento Parcial': 'PagamentoParcial',
    'Flag √Ågio/Des√°gio': '_FlagDesagio'
}
dimensoes_analise = {k: v for k, v in dimensoes_analise.items() if v in df_report.columns} #? garante que s√≥ usamos colunas que existem
os.makedirs(output_path, exist_ok=True) #* cria a pasta de sa√≠da se ela n√£o existir

#***********************
#* C√ÅLCULO M√âTRICAS
#***********************
print("\n" + "="*80)
print("INICIANDO C√ÅLCULO UNIFICADO DAS M√âTRICAS")
print("="*80)

#* cria colunas espec√≠ficas para valor vencido em diferentes faixas de atraso
dias_de_atraso = (data_referencia_relatorio - df_report['DataVencimento']).dt.days
df_report['_ValorVencido_1d'] = df_report.loc[dias_de_atraso >= 1, 'ValorPresente'].fillna(0)
df_report['_ValorVencido_30d'] = df_report.loc[dias_de_atraso >= 30, 'ValorPresente'].fillna(0)
df_report['_ValorVencido_60d'] = df_report.loc[dias_de_atraso >= 60, 'ValorPresente'].fillna(0)
df_report.fillna({'_ValorVencido_1d': 0, '_ValorVencido_30d': 0, '_ValorVencido_60d': 0}, inplace=True) #? preenche nulos com 0

"""vp_col_name, vl_col_name, col_contratos_venc_perc: str
Nomes das colunas que aparecer√£o no relat√≥rio HTML. O uso de \n
permite a quebra de linha no cabe√ßalho da tabela.
"""
vp_col_name = 'Valor Presente \n(R$ MM)'
vl_col_name = 'Valor L√≠quido \n(R$ MM)'
col_contratos_venc_perc = f"% Contratos Venc. (>{DIAS_ATRASO_DEFINICAO_VENCIDO}d)"

"""tabelas_metricas: dict
Dicion√°rio para armazenar os DataFrames de m√©tricas para cada dimens√£o analisada.
"""
tabelas_metricas = {}

#? Itera sobre cada dimens√£o para criar as tabelas de m√©tricas segmentadas
for nome_analise, coluna in dimensoes_analise.items():
    print(f"Calculando m√©tricas para a dimens√£o: '{nome_analise}'...")
    grouped = df_report.groupby(coluna, observed=True) #* agrupa os dados pela coluna da dimens√£o
    
    total_contratos_unicos = grouped['CCB'].nunique() #* conta contratos √∫nicos por segmento
    contratos_vencidos_unicos = df_report[df_report['_ContratoVencido_Flag'] == 1].groupby(coluna, observed=True)['CCB'].nunique()
    
    df_metricas = pd.DataFrame({'N¬∫ Contratos √önicos': total_contratos_unicos}) #? inicia o dataframe de m√©tricas
    somas_financeiras = grouped[['_ValorLiquido', 'ValorPresente', '_ValorVencido_1d', '_ValorVencido_30d', '_ValorVencido_60d']].sum() #* calcula as somas dos valores financeiros
    df_metricas = df_metricas.join(somas_financeiras) #? junta as somas ao dataframe de m√©tricas
    
    #* Calcula as m√©tricas percentuais
    df_metricas['%PDD'] = (1 - df_metricas['_ValorLiquido'] / df_metricas['ValorPresente']) * 100
    df_metricas[col_contratos_venc_perc] = (contratos_vencidos_unicos.reindex(df_metricas.index).fillna(0) / df_metricas['N¬∫ Contratos √önicos']).fillna(0) * 100
    df_metricas['Venc. 1d / VP'] = (df_metricas['_ValorVencido_1d'] / df_metricas['ValorPresente']) * 100
    df_metricas['Venc. 30d / VP'] = (df_metricas['_ValorVencido_30d'] / df_metricas['ValorPresente']) * 100
    df_metricas['Venc. 60d / VP'] = (df_metricas['_ValorVencido_60d'] / df_metricas['ValorPresente']) * 100
    
    #* Renomeia e formata as colunas para o relat√≥rio
    df_metricas = df_metricas.rename(columns={'ValorPresente': vp_col_name, '_ValorLiquido': vl_col_name, 'N¬∫ Contratos √önicos': 'N¬∫ Contratos'})
    df_metricas[[vp_col_name, vl_col_name]] /= 1e6 #? converte para milh√µes
    df_metricas = df_metricas.drop(columns=['_ValorLiquido', '_ValorVencido_1d', '_ValorVencido_30d', '_ValorVencido_60d'], errors='ignore')
    
    tabelas_metricas[nome_analise] = df_metricas #* armazena a tabela final no dicion√°rio
print("C√°lculo unificado de m√©tricas conclu√≠do.")

#***********************
#* TICKET M√âDIO
#***********************
"""tabelas_ticket: dict
Dicion√°rio para armazenar os DataFrames de ticket m√©dio para cada dimens√£o.
"""
tabelas_ticket = {}
for nome_analise, coluna in dimensoes_analise.items():
    df_temp = df_report.dropna(subset=[coluna, 'ValorPresente', 'ValorNominal']) #? ignora linhas com dados faltantes
    if df_temp.empty: continue
    
    grouped = df_temp.groupby(coluna, observed=True)
    #* Calcula o Ticket M√©dio ponderado pelo Valor Presente para dar mais peso √†s parcelas mais valiosas
    numerador = grouped.apply(lambda g: (g['ValorNominal'] * g['ValorPresente']).sum(), include_groups=False)
    denominador = grouped['ValorPresente'].sum()
    ticket_ponderado = (numerador / denominador).replace([np.inf, -np.inf], 0)
    ticket_ponderado.name = "Ticket Ponderado (R$)"
    
    tabelas_ticket[nome_analise] = pd.DataFrame(ticket_ponderado)
print("Ticket m√©dio calculado.")


# << NOVO BLOCO: C√ÅLCULO DAS M√âTRICAS ADICIONAIS ANTES DO LOOP PRINCIPAL >>
#***********************
#* REPRESENTATIVIDADE DA TIR
#***********************
print("Calculando representatividade da amostra para a TIR...")
"""tabelas_representatividade: dict
Armazena, para cada segmento, quantos contratos t√™m parcelas futuras e qual 
percentual isso representa do total de contratos daquele segmento.
"""
tabelas_representatividade = {}
#? Filtra apenas as parcelas que vencem no futuro, que s√£o a base para o c√°lculo da TIR
df_avencer_temp = df_report[df_report['DataVencimento'] >= data_referencia_relatorio]

for nome_analise, coluna in dimensoes_analise.items():
    if coluna not in df_report.columns: continue
    
    total_contratos = df_report.groupby(coluna, observed=True)['CCB'].nunique() #* total de contratos no segmento
    contratos_com_futuro = df_avencer_temp.groupby(coluna, observed=True)['CCB'].nunique() #* total de contratos com parcelas futuras
    
    df_rep = pd.DataFrame({'Total Contratos': total_contratos, 'Amostra TIR (Contratos)': contratos_com_futuro}).fillna(0)
    df_rep['Amostra TIR (Contratos)'] = df_rep['Amostra TIR (Contratos)'].astype(int)
    
    #* Calcula e formata o percentual de representatividade
    percentual = (df_rep['Amostra TIR (Contratos)'] / df_rep['Total Contratos'] * 100).fillna(0)
    df_rep['Amostra TIR (Contratos)'] = df_rep['Amostra TIR (Contratos)'].astype(str) + ' (' + percentual.map('{:,.2f}%'.format) + ')'
    
    tabelas_representatividade[nome_analise] = df_rep[['Amostra TIR (Contratos)']] #? armazena a coluna formatada
print("C√°lculo de representatividade conclu√≠do.")

#***********************
#* DATA MEDIANA DE VENCIMENTO
#***********************
print("Calculando a data mediana de vencimento...")
"""tabelas_data_mediana: dict
Armazena a data de vencimento mediana para as parcelas futuras de cada segmento.
Isso d√° uma ideia do horizonte de tempo da carteira.
"""
tabelas_data_mediana = {}

for nome_analise, coluna in dimensoes_analise.items():
    if coluna not in df_report.columns: continue
    
    medianas = df_avencer_temp.groupby(coluna, observed=True)['DataVencimento'].median() #* calcula a mediana das datas
    df_medianas = pd.DataFrame(medianas)
    df_medianas['DataVencimento'] = df_medianas['DataVencimento'].dt.strftime('%d/%m/%Y').fillna('-') #* formata a data
    df_medianas = df_medianas.rename(columns={'DataVencimento': 'Data Mediana Venc.'})
    
    tabelas_data_mediana[nome_analise] = df_medianas
print("C√°lculo da data mediana conclu√≠do.")


"""#### <span style="color:#FFDAC1;"> C√°lculo da TIR

C√°lculo da TIR (XIRR) para a carteira a vencer. Como n√£o h√° dados de custo, calculamos a TIR Bruta e a TIR L√≠quida de PDD.
"""
#******************
#* TIR com brentq
#*******************
def calculate_xirr(cash_flows, days):
    """
    Calcula a Taxa Interna de Retorno (TIR) para uma s√©rie de fluxos de caixa
    n√£o peri√≥dicos (equivalente ao XIRR do Excel).

    Args:
        cash_flows (list or np.array): Lista de fluxos de caixa. O primeiro
                                    valor deve ser o investimento (negativo).
        days (list or np.array): Lista dos dias correspondentes a cada fluxo,
                                onde o dia 0 √© a data do investimento.

    Returns:
        float: A TIR mensal, ou np.nan se n√£o for poss√≠vel calcular.
    """
    cash_flows = np.array(cash_flows)
    days = np.array(days)
    
    def npv(rate): #? fun√ß√£o interna de Valor Presente L√≠quido (VPL) que ser√° zerada
        if rate <= -1: return float('inf') #! previne erros matem√°ticos com taxas <= -100%
        with np.errstate(divide='ignore', over='ignore'):
            #* a f√≥rmula do VPL: soma de cada fluxo dividido por (1+taxa) elevado aos dias
            #* a base de dias √© 21 para anualizar a taxa mensal para dias √∫teis.
            return np.sum(cash_flows / (1 + rate) ** (days / 21.0))
    try:
        #? usa o m√©todo de brentq para encontrar a raiz (a taxa 'rate') que torna npv(rate) = 0
        return brentq(npv, 0, 1.0)
    except (RuntimeError, ValueError): #! se n√£o encontrar uma solu√ß√£o no intervalo, retorna NaN
        return np.nan

print("\n" + "="*80)
print("INICIANDO C√ÅLCULO DA TAXA INTERNA DE RETORNO (TIR)")
print("="*80)

ref_date = df_report['DataGeracao'].max()
print(f"Data de Refer√™ncia para o c√°lculo da TIR: {ref_date.strftime('%d/%m/%Y')}")

try:
    df_feriados = pd.read_excel(caminho_feriados)
    holidays = pd.to_datetime(df_feriados['Data'], errors='coerce').dropna().dt.date.values
    print(f"Sucesso: {len(holidays)} feriados carregados.")
except Exception as e:
    print(f"[AVISO] N√£o foi poss√≠vel carregar feriados: {e}. O c√°lculo de dias √∫teis pode ser impreciso.")
    holidays = []

#? df_avencer √© a base de dados para o c√°lculo da TIR, contendo apenas parcelas futuras.
df_avencer = df_report[df_report['DataVencimento'] >= ref_date].copy()

if not df_avencer.empty:
    try:
        start_date_np = np.datetime64(ref_date.date(), 'D') #? data inicial para contagem de dias
        end_dates_np = df_avencer['DataVencimento'].dt.date.values.astype('datetime64[D]')
        holidays_np = np.array(holidays, dtype='datetime64[D]')

        #* calcula o n√∫mero de dias √∫teis entre a data de refer√™ncia e o vencimento de cada parcela
        df_avencer.loc[:, '_DIAS_UTEIS_'] = np.busday_count(start_date_np, end_dates_np, holidays=holidays_np)
        
        print("C√°lculo de dias √∫teis realizado com sucesso.")
    except Exception as e:
        print(f"[ERRO] Falha ao calcular dias √∫teis: {e}")
        df_avencer.loc[:, '_DIAS_UTEIS_'] = np.nan
else:
    print("[AVISO] N√£o h√° parcelas futuras ('a vencer') na base de dados. O c√°lculo de TIR ser√° pulado.")
    df_avencer['_DIAS_UTEIS_'] = np.nan

"""all_tirs: list
Lista para armazenar os dicion√°rios com os resultados da TIR para cada segmento.
"""
all_tirs = []
cat_cols = list(dimensoes_analise.values())
#? cria uma lista de todos os segmentos que precisam ter a TIR calculada
segmentos_para_analise = [('Carteira Total', 'Todos')] + \
                         [(col, seg) for col in cat_cols if col in df_avencer.columns for seg in df_avencer[col].dropna().unique()]

for tipo_dimensao, segmento in segmentos_para_analise:
    if tipo_dimensao == 'Carteira Total':
        df_segmento = df_avencer.copy()
    else:
        df_segmento = df_avencer[df_avencer[tipo_dimensao] == segmento].copy()

    if df_segmento.empty or df_segmento['_DIAS_UTEIS_'].isnull().all():
        continue

    #* o 'investimento' √© a soma do Valor Presente de todas as parcelas futuras do segmento
    vp_bruto = df_segmento['ValorPresente'].sum()
    tir_bruta, tir_pdd = np.nan, np.nan

    if vp_bruto > 0:
        # --- TIR Bruta ---
        fluxos_brutos = df_segmento.groupby('_DIAS_UTEIS_')['ValorNominal'].sum() #* agrupa os fluxos de caixa por dia
        tir_bruta = calculate_xirr([-vp_bruto] + fluxos_brutos.values.tolist(), [0] + fluxos_brutos.index.tolist())

        # --- TIR L√≠quida de PDD ---
        pdd_total_segmento = df_segmento['PDDTotal'].sum()
        pdd_rate = (pdd_total_segmento / vp_bruto) if pd.notna(pdd_total_segmento) and vp_bruto > 0 else 0.0
        df_segmento['Fluxo_PDD'] = df_segmento['ValorNominal'] * (1 - pdd_rate) #* ajusta o fluxo de caixa pela taxa de PDD
        fluxos_pdd = df_segmento.groupby('_DIAS_UTEIS_')['Fluxo_PDD'].sum()
        tir_pdd = calculate_xirr([-vp_bruto] + fluxos_pdd.values.tolist(), [0] + fluxos_pdd.index.tolist())

    all_tirs.append({
        'DimensaoColuna': tipo_dimensao, 'Segmento': segmento,
        'Valor Presente TIR (M)': vp_bruto / 1e6,
        'TIR Bruta \n(% a.m. )': tir_bruta * 100 if pd.notna(tir_bruta) else np.nan,
        'TIR L√≠quida de PDD \n(% a.m. )': tir_pdd * 100 if pd.notna(tir_pdd) else np.nan
    })

df_tir_summary = pd.DataFrame(all_tirs)
if not df_tir_summary.empty:
    #? Se o c√°lculo da TIR falhar (resultando em NaN), preenche com -100% para indicar o erro no relat√≥rio
    tir_cols_to_fill = [col for col in df_tir_summary.columns if 'TIR' in col]
    df_tir_summary[tir_cols_to_fill] = df_tir_summary[tir_cols_to_fill].fillna(-100.0)

print("C√°lculo de TIR conclu√≠do.")

"""#### <span style="color:#FFDAC1;"> Montagem do HTML

Uno todos os elementos em um relat√≥rio HTML interativo.
"""
#***********************
#* GERA√á√ÉO DO RELAT√ìRIO HTML
#***********************

print("\n" + "="*80)
print("GERANDO RELAT√ìRIO HTML FINAL")
print("="*80)

def encode_image_to_base64(image_path):
    """Codifica uma imagem para o formato base64 para embuti-la no HTML."""
    try:
        with open(image_path, "rb") as image_file:
            return base64.b64encode(image_file.read()).decode('utf-8')
    except FileNotFoundError:
        print(f"[ATEN√á√ÉO] Arquivo de imagem n√£o encontrado em: {image_path}. A logo n√£o ser√° exibida.")
        return None

logo_base64 = encode_image_to_base64(logo_path)
report_date = ref_date.strftime('%d/%m/%Y')

"""html_css: str
String contendo todo o c√≥digo CSS para estilizar o relat√≥rio final.
"""
html_css = """
<style>
    .single-entry-message { background-color: #eef7f7; border-left: 5px solid #76c6c5; padding: 15px 20px; margin: 10px 0; font-size: 1.0em; color: #313131; }
    .single-entry-message strong { color: #163f3f; }
    .summary-title { font-size: 1.4em; color: #163f3f; border-bottom: 2px solid #76c6c5; padding-bottom: 10px; margin-top: 20px; margin-bottom: 20px; }
    .checks-container { display: grid; grid-template-columns: repeat(auto-fit, minmax(350px, 1fr)); gap: 15px; margin-bottom: 20px; }
    .check-item { background-color: #f5f5f5; padding: 12px; border-radius: 5px; border-left: 4px solid #76c6c5; font-size: 0.95em; }
    .check-item strong { color: #163f3f; }
    .dataframe-checks th { background-color: #5b8c8c; }
    .check-table-wrapper h4 { margin-top: 20px; margin-bottom: 10px; color: #163f3f; border-bottom: 2px solid #eeeeee; padding-bottom: 5px; }
    body { font-family: "Gill Sans MT", Arial, sans-serif; background-color: #f9f9f9; color: #313131; margin: 0; padding: 0; }
    .main-content { padding: 25px; }
    header { background-color: #163f3f; color: #FFFFFF; padding: 20px 40px; display: flex; justify-content: space-between; align-items: center; border-bottom: 5px solid #76c6c5; }
    header .logo img { height: 75px; }
    header .report-title { text-align: left; font-family: "Gill Sans MT", Arial, sans-serif; }
    header .report-title h1, header .report-title h2, header .report-title h3 { margin: 0; padding: 0; font-weight: normal; }
    header .report-title h1 { font-size: 1.6em; }
    header .report-title h2 { font-size: 1.4em; color: #d0d0d0; }
    header .report-title h3 { font-size: 1.1em; color: #a0a0a0; }
    .container-botoes { display: flex; flex-wrap: wrap; gap: 15px; margin-bottom: 25px; margin-top: 40px; }
    .container-botoes > details { flex: 1 1 280px; border: 1px solid #76c6c5; border-radius: 8px; overflow: hidden; }
    .container-botoes > details[open] { flex-basis: 100%; }
    details summary { font-size: 1.1em; font-weight: bold; color: #FFFFFF; background-color: #163f3f; padding: 15px 20px; cursor: pointer; outline: none; list-style-type: none; }
    details summary:hover { background-color: #0e5d5f; }
    details[open] summary { background-color: #76c6c5; color: #313131; }
    details[open] summary:hover { filter: brightness(95%); }
    summary::-webkit-details-marker { display: none; }
    summary::before { content: '‚ñ∫ '; margin-right: 8px; font-size: 0.8em;}
    details[open] summary::before { content: '‚ñº '; }
    .content-wrapper { padding: 20px; background-color: #FFFFFF; overflow-x: auto; border: 1px solid #ddd; border-radius: 8px; }
    table.dataframe, th, td { border: 1px solid #bbbbbb; }
    table.dataframe { border-collapse: collapse; width: 100%; }
    th, td { text-align: left; padding: 10px; vertical-align: middle; }
    th { background-color: #163f3f; color: #FFFFFF; }
    tr:nth-child(even) { background-color: #eeeeee; }
    footer { background-color: #f0f0f0; color: #555555; font-size: 0.8em; line-height: 1.6; padding: 25px 40px; margin-top: 40px; border-top: 1px solid #dddddd; }
    footer .disclaimer { margin-top: 20px; font-style: italic; border-top: 1px solid #dddddd; padding-top: 20px; }
</style>
"""

#* Monta a estrutura principal do HTML, come√ßando pelo cabe√ßalho
html_parts = ["<!DOCTYPE html><html lang='pt-BR'><head>",
              "<meta charset='UTF-8'><title>An√°lise de Desempenho - FCT Consignado II</title>",
              html_css, "</head><body>"]

html_parts.append(f"<header><div class='report-title'><h1>An√°lise de desempenho</h1><h2>FCT CONSIGNADO II</h2><h3>{report_date}</h3></div>")
if logo_base64:
    html_parts.append(f'<div class="logo"><img src="data:image/png;base64,{logo_base64}" alt="Logo"></div>')
html_parts.append("</header><div class='main-content'>")

"""mapa_descricoes: dict
Dicion√°rio para adicionar dicas (tooltips) aos t√≠tulos das se√ß√µes no HTML.
"""
mapa_descricoes = {
    'Cedentes': 'Analisa as m√©tricas de risco e retorno agrupadas por cada Cedente.',
    'Tipo de Ativo': 'Compara o desempenho entre os diferentes tipos de ativos (CCB, CT - Contrato).',
    'Situa√ß√£o': 'Compara o desempenho dos t√≠tulos com base na sua situa√ß√£o atual (Ex: Aditado).',
    'Pagamento Parcial': 'Verifica se h√° impacto nas m√©tricas para t√≠tulos que aceitam pagamento parcial.',
    'Flag √Ågio/Des√°gio': 'Compara o desempenho de contratos adquiridos com √°gio vs. des√°gio.'
}

#* Adiciona a se√ß√£o de Verifica√ß√µes de Sanidade ao HTML
html_parts.append("<details open><summary>Verifica√ß√µes e Sanidade dos Dados</summary><div class='content-wrapper'>")
simple_checks_html = "<div class='checks-container'>"
table_checks_html = "<div class='check-table-wrapper'>"
for key, value in checks_results.items():
    if '[TABELA]' in key:
        clean_key = key.replace('[TABELA]', '').strip()
        table_checks_html += f"<h4>{clean_key}</h4>{str(value)}"
    else:
        simple_checks_html += f"<div class='check-item'><strong>{key}:</strong> {value}</div>"
simple_checks_html += "</div>"
table_checks_html += "</div>"
html_parts.append(simple_checks_html)
html_parts.append(table_checks_html)
html_parts.append("</div></details>")

# << ALTERADO: Bloco do Resumo da Carteira Total agora tamb√©m inclui as novas m√©tricas >>
# Resumo da Carteira Total
print("--> Gerando tabela de resumo da carteira total...")
df_avencer_total = df_report[df_report['DataVencimento'] >= data_referencia_relatorio]
data_mediana_total = df_avencer_total['DataVencimento'].median()

total_metrics = {
    'N¬∫ Contratos': df_report['CCB'].nunique(),
    'Amostra TIR (Contratos)': f"{df_avencer_total['CCB'].nunique():,} ({df_avencer_total['CCB'].nunique() / df_report['CCB'].nunique() * 100:.2f}%)",
    'Data Mediana Venc.': data_mediana_total.strftime('%d/%m/%Y') if pd.notna(data_mediana_total) else '-',
    col_contratos_venc_perc: (df_report[df_report['_ContratoVencido_Flag'] == 1]['CCB'].nunique() / df_report['CCB'].nunique()) * 100,
    vl_col_name: df_report['_ValorLiquido'].sum() / 1e6,
    vp_col_name: df_report['ValorPresente'].sum() / 1e6,
    '%PDD': (1 - df_report['_ValorLiquido'].sum() / df_report['ValorPresente'].sum()) * 100 if df_report['ValorPresente'].sum() > 0 else 0,
    'Venc. 1d / VP': (df_report['_ValorVencido_1d'].sum() / df_report['ValorPresente'].sum()) * 100 if df_report['ValorPresente'].sum() > 0 else 0,
    'Venc. 30d / VP': (df_report['_ValorVencido_30d'].sum() / df_report['ValorPresente'].sum()) * 100 if df_report['ValorPresente'].sum() > 0 else 0,
    'Venc. 60d / VP': (df_report['_ValorVencido_60d'].sum() / df_report['ValorPresente'].sum()) * 100 if df_report['ValorPresente'].sum() > 0 else 0,
    'Ticket Ponderado (R$)': (df_report['ValorNominal'] * df_report['ValorPresente']).sum() / df_report['ValorPresente'].sum() if df_report['ValorPresente'].sum() > 0 else 0,
}
df_total = pd.DataFrame([total_metrics], index=['Total'])
if not df_tir_summary.empty:
    df_tir_total = df_tir_summary[df_tir_summary['DimensaoColuna'] == 'Carteira Total'].drop(columns=['DimensaoColuna', 'Segmento', 'Valor Presente TIR (M)'])
    if not df_tir_total.empty:
        df_tir_total.index = ['Total']
        df_total = pd.concat([df_total, df_tir_total], axis=1)

# << ALTERADO: Ordem das colunas do resumo total >>
colunas_vencimento = ['Venc. 1d / VP', 'Venc. 30d / VP', 'Venc. 60d / VP']
colunas_ordem_total = ['N¬∫ Contratos', 'Amostra TIR (Contratos)', 'Data Mediana Venc.', col_contratos_venc_perc, vl_col_name, vp_col_name, '%PDD'] + colunas_vencimento + ['Ticket Ponderado (R$)']
ordem_ideal_tir = ['TIR Bruta \n(% a.m. )', 'TIR L√≠quida de PDD \n(% a.m. )']
colunas_tir_ordenadas = [col for col in ordem_ideal_tir if col in df_total.columns]
colunas_finais_total = [col for col in colunas_ordem_total + colunas_tir_ordenadas if col in df_total.columns]
df_total = df_total[colunas_finais_total]

#? dicion√°rio para formatar as colunas da tabela
formatters = {
    vl_col_name: '{:,.2f}'.format, vp_col_name: '{:,.2f}'.format,
    'N¬∫ Contratos': lambda x: f'{x:,.0f}'.replace(',', '.'),
    'Ticket Ponderado (R$)': lambda x: f'R$ {x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.'),
    '%PDD': '{:,.2f}%'.format
}
formatters[col_contratos_venc_perc] = '{:,.2f}%'.format
for col in colunas_vencimento: formatters[col] = '{:,.2f}%'.format
for col in colunas_tir_ordenadas: formatters[col] = '{:,.2f}%'.format

df_total.columns = [col.replace('\n', '<br>') for col in df_total.columns] #? substitui \n por <br> para HTML

#* adiciona a tabela de resumo total ao HTML
html_parts.append("<h2 class='summary-title'>Resumo da Carteira Total</h2><div class='content-wrapper'>")
html_parts.append(df_total.to_html(index=False, classes='dataframe', formatters=formatters, na_rep='-', escape=False))
html_parts.append("</div>")


# << ALTERADO: Loop principal agora junta as novas tabelas e exibe as novas colunas >>
# In√≠cio dos bot√µes retr√°teis (se√ß√µes sanfonadas)
html_parts.append("<div class='container-botoes'>")
for nome_analise, coluna in dimensoes_analise.items():
    if coluna not in df_report.columns or df_report[coluna].isnull().all(): continue
    print(f"--> Processando e gerando HTML para: '{nome_analise}'")

    df_final = tabelas_metricas.get(nome_analise)
    df_ticket = tabelas_ticket.get(nome_analise, pd.DataFrame())
    df_rep = tabelas_representatividade.get(nome_analise, pd.DataFrame())
    df_mediana = tabelas_data_mediana.get(nome_analise, pd.DataFrame())

    #* Juntando todas as tabelas calculadas (m√©tricas, representatividade, mediana, ticket e TIR)
    df_final = df_final.join(df_rep, how='outer')
    df_final = df_final.join(df_mediana, how='outer')
    df_final = df_final.join(df_ticket, how='outer')
    if not df_tir_summary.empty:
        df_tir = df_tir_summary[df_tir_summary['DimensaoColuna'] == coluna].set_index('Segmento')
        if not df_tir.empty:
            df_final = df_final.join(df_tir.drop(columns=['DimensaoColuna', 'Valor Presente TIR (M)']), how='outer')

    df_final.index.name = nome_analise
    df_final.reset_index(inplace=True)

    #* Define a ordem final das colunas na tabela HTML
    colunas_ordem = [
        nome_analise, 'N¬∫ Contratos', 'Amostra TIR (Contratos)', 'Data Mediana Venc.', 
        col_contratos_venc_perc, vl_col_name, vp_col_name, '%PDD'
    ] + colunas_vencimento
    
    if 'Ticket Ponderado (R$)' in df_final.columns: colunas_ordem.append('Ticket Ponderado (R$)')

    colunas_finais = [col for col in colunas_ordem + colunas_tir_ordenadas if col in df_final.columns]
    outras_colunas = [col for col in df_final.columns if col not in df_final.columns]
    df_final = df_final[colunas_finais + outras_colunas]

    #* Ordena as linhas da tabela pelo Valor Presente (maior para o menor)
    df_final = df_final.sort_values(vp_col_name, ascending=False).reset_index(drop=True)
    df_final.columns = [col.replace('\n', '<br>') for col in df_final.columns]

    #* Cria a se√ß√£o sanfonada (<details>) para cada an√°lise
    html_parts.append("<details>")
    descricao = mapa_descricoes.get(nome_analise, 'Descri√ß√£o n√£o dispon√≠vel.')
    html_parts.append(f'<summary title="{descricao}">{nome_analise}</summary>')
    html_parts.append("<div class='content-wrapper'>")
    html_table = df_final.to_html(index=False, classes='dataframe', formatters=formatters, na_rep='-', escape=False)
    html_parts.append(html_table)
    html_parts.append("</div></details>")

html_parts.append("</div></div>") #? fecha as divs do container e main-content

#* Adiciona o rodap√© com o disclaimer
footer_main_text = "Este documento tem como objetivo apresentar uma an√°lise de desempenho do fundo FCT Consignado II (CNPJ 52.203.615/0001-19)..."
footer_disclaimer = "Disclaimer: Este relat√≥rio foi preparado pelo Porto Real Investimentos exclusivamente para fins informativos..."
html_parts.append(f"<footer><p>{footer_main_text.strip()}</p><div class='disclaimer'>{footer_disclaimer.strip()}</div></footer>")
html_parts.append("</body></html>")

#* Junta todas as partes do HTML em uma √∫nica string e salva em um arquivo
final_html_content = "\n".join(html_parts)
html_output_filename = os.path.join(output_path, 'analise_carteira_estoque.html')
try:
    with open(html_output_filename, 'w', encoding='utf-8') as f:
        f.write(final_html_content)
    print("\n" + "="*80)
    print("AN√ÅLISE CONCLU√çDA COM SUCESSO!")
    print(f"O relat√≥rio HTML final foi salvo em: {html_output_filename}")
    print("="*80)
except Exception as e:
    print(f"\n[ERRO GRAVE] N√£o foi poss√≠vel salvar o arquivo HTML: {e}")

#del df_report, df_total
```
